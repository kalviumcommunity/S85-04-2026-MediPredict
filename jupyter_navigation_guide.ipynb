## 🏠 Jupyter Home Interface Components

### Key Areas to Identify:
1. **File Browser**: Lists files and folders in current directory
2. **Navigation Breadcrumbs**: Shows current path and allows navigation
3. **New Button**: Dropdown menu for creating new files/notebooks
4. **Upload Button**: For uploading files to current directory
5. **Refresh Button**: Reload the file browser
6. **Running Tab**: Shows active kernels and sessions
7. **File Type Indicators**: Icons for folders, notebooks, scripts, etc.

### File Type Icons:
- 📁 **Folder**: Directory
- 📓 **Notebook**: .ipynb files
- 📄 **File**: Other file types (.py, .txt, .csv, etc.)

````python
# Let's explore the current directory structure
import os

def explore_directory(path=".", max_depth=2):
    """Explore directory structure with depth limit"""
    for root, dirs, files in os.walk(path):
        depth = root.count(os.sep) - path.count(os.sep)
        if depth <= max_depth:
            level = "  " * depth
            print(f"{level}📁 {os.path.basename(root)}/")
            sublevel = "  " * (depth + 1)
            for file in files[:5]:  # Limit to first 5 files
                if file.endswith(('.ipynb', '.py', '.md', '.txt')):
                    print(f"{sublevel}📄 {file}")
            if len(files) > 5:
                print(f"{sublevel}... and {len(files)-5} more files")

print("📂 Current Directory Structure:")
explore_directory()
print("\n✅ Directory exploration complete!")
````

## 📓 Notebook Management Basics

### Essential Operations:
1. **Create New Notebook**: New → Python 3
2. **Rename Notebook**: File → Rename or click title
3. **Save Notebook**: Ctrl+S or floppy disk icon
4. **Close Notebook**: File → Close and Halt
5. **Reopen Notebook**: Click .ipynb file in browser

### Keyboard Shortcuts:
- **Ctrl+S**: Save notebook
- **Ctrl+Enter**: Run current cell
- **Shift+Enter**: Run cell and select below
- **Alt+Enter**: Run cell and insert below
- **DD**: Delete current cell (in command mode)
- **A**: Insert cell above (command mode)
- **B**: Insert cell below (command mode)
- **M**: Change to markdown (command mode)
- **Y**: Change to code (command mode)
