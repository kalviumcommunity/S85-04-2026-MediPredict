# Project Structure Verification

This notebook verifies that the MediPredict project structure is properly organized according to Data Science best practices.

## Purpose
- Verify folder structure exists and is accessible
- Check that all required directories are present
- Demonstrate proper file path usage
- Validate organization for collaboration

**Environment**: datascience conda environment  
**Project**: MediPredict Healthcare Forecasting

```python
# Import required libraries
import os
import pandas as pd
from pathlib import Path

print("🔍 Project Structure Verification Started")
print(f"📍 Current working directory: {os.getcwd()}")
print(f"🐍 Python version: {pd.__version__}")
```

## 1. Verify Required Folders Exist

Let's check that all the essential folders for a Data Science project are present and accessible.

```python
# Define required folder structure
required_folders = {
    'data': ['raw', 'processed', 'external'],
    'notebooks': ['eda', 'modeling', 'visualization', 'reports', 'verification'],
    'scripts': ['data_processing', 'modeling', 'utilities'],
    'outputs': ['figures', 'reports'],
    'models': [],
    'docs': [],
    'backend': [],
    'frontend': [],
    'proof': []
}

# Verify folder existence
print("📁 Verifying folder structure:")
all_folders_exist = True

for main_folder, subfolders in required_folders.items():
    main_path = Path(main_folder)
    
    if main_path.exists():
        print(f"  ✅ {main_folder}/ - EXISTS")
        
        # Check subfolders
        for subfolder in subfolders:
            sub_path = main_path / subfolder
            if sub_path.exists():
                print(f"    ✅ {subfolder}/ - EXISTS")
            else:
                print(f"    ❌ {subfolder}/ - MISSING")
                all_folders_exist = False
    else:
        print(f"  ❌ {main_folder}/ - MISSING")
        all_folders_exist = False

print(f"\n🎯 Overall folder status: {'✅ COMPLETE' if all_folders_exist else '❌ INCOMPLETE'}")
```

## 2. Verify Data Organization

Check that data folders are properly structured and contain appropriate files.

```python
# Check data folder structure
data_folders = ['data/raw', 'data/processed', 'data/external']

print("📊 Data folder verification:")
for folder in data_folders:
    path = Path(folder)
    if path.exists():
        files = list(path.glob('*'))
        print(f"  ✅ {folder}/ - {len(files)} files")
        for file in files[:3]:  # Show first 3 files
            print(f"    📄 {file.name}")
        if len(files) > 3:
            print(f"    ... and {len(files)-3} more files")
    else:
        print(f"  ❌ {folder}/ - MISSING")
```

## 3. Verify Documentation Structure

Check that documentation files are present and accessible.

```python
# Check documentation files
docs_folder = Path('docs')
required_docs = ['PROJECT_STRUCTURE.md', 'DATA_DICTIONARY.md']

print("📚 Documentation verification:")
if docs_folder.exists():
    print(f"  ✅ docs/ - EXISTS")
    
    for doc_file in required_docs:
        doc_path = docs_folder / doc_file
        if doc_path.exists():
            size = doc_path.stat().st_size
            print(f"    ✅ {doc_file} - {size:,} bytes")
        else:
            print(f"    ❌ {doc_file} - MISSING")
else:
    print(f"  ❌ docs/ - MISSING")

# Check main README
readme_path = Path('README.md')
if readme_path.exists():
    size = readme_path.stat().st_size
    print(f"  ✅ README.md - {size:,} bytes")
else:
    print(f"  ❌ README.md - MISSING")
```

## 4. Verify Notebook Organization

Check that notebooks are properly categorized in their respective folders.

```python
# Check notebook organization
notebook_folders = ['notebooks/eda', 'notebooks/modeling', 'notebooks/visualization', 
                   'notebooks/reports', 'notebooks/verification']

print("📓 Notebook organization verification:")
total_notebooks = 0

for folder in notebook_folders:
    path = Path(folder)
    if path.exists():
        notebooks = list(path.glob('*.ipynb'))
        total_notebooks += len(notebooks)
        print(f"  ✅ {folder}/ - {len(notebooks)} notebooks")
        
        for notebook in notebooks[:2]:  # Show first 2 notebooks
            print(f"    📓 {notebook.name}")
        if len(notebooks) > 2:
            print(f"    ... and {len(notebooks)-2} more notebooks")
    else:
        print(f"  ❌ {folder}/ - MISSING")

print(f"\n📊 Total notebooks found: {total_notebooks}")
```

## 5. Test File Path Operations

Demonstrate that file paths work correctly with the new structure.

```python
# Test file path operations
print("🔍 Testing file path operations:")

# Test relative paths
data_path = Path('data/raw')
notebooks_path = Path('notebooks/eda')
scripts_path = Path('scripts/data_processing')

paths_to_test = [
    ('Data raw folder', data_path),
    ('EDA notebooks folder', notebooks_path),
    ('Data processing scripts', scripts_path)
]

for description, path in paths_to_test:
    if path.exists():
        absolute_path = path.resolve()
        print(f"  ✅ {description}: {absolute_path}")
    else:
        print(f"  ❌ {description}: Path not found")

# Test creating a sample file path
sample_data_path = Path('data/processed/sample_processed.csv')
print(f"\n📝 Sample processed data path: {sample_data_path}")
print(f"📁 Parent directory: {sample_data_path.parent}")
print(f"📄 File name: {sample_data_path.name}")
print(f"🏷️  File stem: {sample_data_path.stem}")
print(f"📂 File suffix: {sample_data_path.suffix}")
```

## 6. Collaboration Readiness Check

Verify that the structure supports collaboration and reproducibility.

```python
# Collaboration readiness checklist
print("🤝 Collaboration readiness checklist:")

collaboration_checks = [
    ('README.md exists', Path('README.md').exists()),
    ('Project structure documented', Path('docs/PROJECT_STRUCTURE.md').exists()),
    ('Data dictionary available', Path('docs/DATA_DICTIONARY.md').exists()),
    ('Data folders separated', all([Path('data/raw').exists(), Path('data/processed').exists()])),
    ('Notebooks categorized', Path('notebooks').exists()),
    ('Scripts organized', Path('scripts').exists()),
    ('Output folders ready', Path('outputs').exists()),
    ('Models folder exists', Path('models').exists())
]

passed_checks = 0
total_checks = len(collaboration_checks)

for check_name, result in collaboration_checks:
    status = "✅ PASS" if result else "❌ FAIL"
    print(f"  {status} {check_name}")
    if result:
        passed_checks += 1

collaboration_score = (passed_checks / total_checks) * 100
print(f"\n🎯 Collaboration readiness: {collaboration_score:.1f}% ({passed_checks}/{total_checks} checks passed)")

if collaboration_score >= 80:
    print("🎉 Project structure is collaboration-ready!")
else:
    print("⚠️  Some improvements needed for better collaboration.")
```

## 7. Summary and Recommendations

Final assessment of the project structure and recommendations for improvement.

```python
# Generate final summary
print("📋 Project Structure Verification Summary")
print("=" * 50)

# Count total directories and files
project_root = Path('.')
all_dirs = [d for d in project_root.rglob('*') if d.is_dir()]
all_files = [f for f in project_root.rglob('*') if f.is_file()]

print(f"📁 Total directories: {len(all_dirs)}")
print(f"📄 Total files: {len(all_files)}")
print(f"📓 Jupyter notebooks: {len([f for f in all_files if f.suffix == '.ipynb'])}")
print(f"🐍 Python scripts: {len([f for f in all_files if f.suffix == '.py'])}")
print(f"📊 Data files: {len([f for f in all_files if f.suffix in ['.csv', '.json', '.xlsx']])}")

print(f"\n🎯 Project Structure Status: {'✅ VERIFIED' if all_folders_exist else '❌ NEEDS ATTENTION'}")

print("\n📝 Recommendations:")
print("1. Keep data files in appropriate data/ subfolders")
print("2. Use notebooks/ for analysis, scripts/ for automation")
print("3. Store generated outputs in outputs/ folder")
print("4. Update documentation when structure changes")
print("5. Follow naming conventions consistently")

print("\n🚀 Project is ready for Data Science work!")
```
