## 1. Basic if Statements - Simple Conditional Logic

### What is an if Statement?
An **if statement** allows your program to execute code only when a specific condition is `True`. This is the foundation of decision-making in Python.

### Key Concepts:
- ✅ **Condition**: An expression that evaluates to `True` or `False`
- ✅ **Indentation**: Python uses indentation (4 spaces) to define code blocks
- ✅ **Colon**: The `:` character follows the condition
- ✅ **Boolean Logic**: Conditions use comparison operators

### Comparison Operators:
- `==` : Equal to
- `!=` : Not equal to
- `<` : Less than
- `>` : Greater than
- `<=` : Less than or equal to
- `>=` : Greater than or equal to

### Common Use Cases:
- Validating data values
- Filtering records based on criteria
- Triggering actions when thresholds are met

```python
# Basic if Statement Examples
print("=== BASIC if STATEMENTS ===")

# Example 1: Checking numeric condition
patient_temperature = 101.5

if patient_temperature > 100.4:
    print(f"⚠️  Patient has fever: {patient_temperature}°F")
    print("🩺 Medical attention required")

print(f"\nTemperature recorded: {patient_temperature}°F")

# Example 2: Checking string condition
patient_status = "Critical"

if patient_status == "Critical":
    print(f"\n🚨 Patient status: {patient_status}")
    print("📞 Notify emergency team immediately")

# Example 3: Multiple independent if statements
heart_rate = 95
blood_pressure = 140

print("\n=== MULTIPLE INDEPENDENT CHECKS ===")

if heart_rate > 100:
    print(f"⚠️  Elevated heart rate: {heart_rate} bpm")

if blood_pressure > 130:
    print(f"⚠️  High blood pressure: {blood_pressure} mmHg")

if heart_rate <= 100 and blood_pressure <= 130:
    print("✅ Vital signs within normal range")
```

```python
# Data Validation with if Statements
print("\n=== DATA VALIDATION EXAMPLES ===")

# Validating patient age
patient_age = 25

if patient_age < 0:
    print(f"❌ Invalid age: {patient_age} - Age cannot be negative")

if patient_age > 120:
    print(f"❌ Invalid age: {patient_age} - Age exceeds human maximum")

if 0 <= patient_age <= 120:
    print(f"✅ Valid patient age: {patient_age} years")

# Checking for missing data
patient_name = ""

if patient_name == "":
    print("\n⚠️  Patient name is missing - please provide name")

patient_name = "John Doe"

if patient_name != "":
    print(f"✅ Patient name recorded: {patient_name}")

# Checking list membership
valid_blood_types = ['A+', 'A-', 'B+', 'B-', 'AB+', 'AB-', 'O+', 'O-']
patient_blood_type = 'O+'

if patient_blood_type in valid_blood_types:
    print(f"\n✅ Valid blood type: {patient_blood_type}")

if patient_blood_type not in valid_blood_types:
    print(f"❌ Invalid blood type: {patient_blood_type}")
```

## 2. if-else Statements - Handling Both Outcomes

### What is if-else?
The **if-else** structure allows you to handle both `True` and `False` outcomes of a condition. When the condition is `False`, the `else` block executes instead.

### Key Concepts:
- ✅ **Two Paths**: Code handles both true and false conditions
- ✅ **Mutually Exclusive**: Only one block executes
- ✅ **Complete Coverage**: No condition goes unhandled
- ✅ **Clear Logic**: Both outcomes are explicitly defined

### Common Use Cases:
- Binary decisions (pass/fail, yes/no, valid/invalid)
- Toggle states or flags
- Simple categorization tasks

```python
# Basic if-else Examples
print("=== if-else DECISION BRANCHING ===")

# Example 1: Simple pass/fail logic
test_score = 78
passing_threshold = 70

print(f"Test score: {test_score}")

if test_score >= passing_threshold:
    print("✅ Result: PASSED")
    print("🎉 Congratulations! You passed the exam.")
else:
    print("❌ Result: FAILED")
    print("📚 Please review the material and retake the exam.")

# Example 2: Medical eligibility check
patient_age = 17
min_age_for_procedure = 18

print(f"\nPatient age: {patient_age}")
print(f"Minimum age required: {min_age_for_procedure}")

if patient_age >= min_age_for_procedure:
    print("✅ Patient is eligible for the procedure")
    print("📋 Proceed with scheduling")
else:
    print("❌ Patient is not eligible")
    years_to_wait = min_age_for_procedure - patient_age
    print(f"⏳ Must wait {years_to_wait} more year(s)")
    print("👨‍👩‍👧‍👦 Guardian consent required")
```

```python
# Practical if-else Applications
print("\n=== PRACTICAL if-else APPLICATIONS ===")

# Example 3: Data quality check
heart_rate = 45

print(f"Heart rate reading: {heart_rate} bpm")

if 60 <= heart_rate <= 100:
    print("✅ Normal heart rate")
    print("📊 Data quality: GOOD")
else:
    print("⚠️  Abnormal heart rate detected")
    if heart_rate < 60:
        print("🩺 Possible bradycardia (slow heart rate)")
    else:
        print("🩺 Possible tachycardia (fast heart rate)")
    print("🔍 Please verify measurement")

# Example 4: Risk category assignment
bmi = 28.5

print(f"\nBMI: {bmi}")

if bmi < 18.5:
    category = "Underweight"
    risk_level = "Low"
elif bmi < 25:
    category = "Normal"
    risk_level = "Low"
elif bmi < 30:
    category = "Overweight"
    risk_level = "Moderate"
else:
    category = "Obese"
    risk_level = "High"

print(f"Category: {category}")
print(f"Risk Level: {risk_level}")

if risk_level == "High":
    print("⚠️  Recommend immediate lifestyle intervention")
else:
    print("✅ Continue monitoring")
```

## 3. if-elif-else Statements - Multiple Conditions

### What is if-elif-else?
The **if-elif-else** structure handles multiple conditions in sequence. Python checks each condition in order and executes the first matching block.

### Key Concepts:
- ✅ **Sequential Checking**: Conditions evaluated top to bottom
- ✅ **First Match Wins**: Only first true condition executes
- ✅ **Mutually Exclusive**: Only one branch executes
- ✅ **Order Matters**: Arrange conditions from specific to general
- ✅ **Optional else**: Catch-all for unmatched conditions

### Common Use Cases:
- Multi-level categorization
- Grading systems
- Risk stratification
- Decision trees

```python
# Basic if-elif-else Examples
print("=== if-elif-else MULTIPLE CONDITIONS ===")

# Example 1: Blood pressure classification
systolic_bp = 145
diastolic_bp = 92

print(f"Blood Pressure: {systolic_bp}/{diastolic_bp} mmHg")

if systolic_bp < 120 and diastolic_bp < 80:
    bp_category = "Normal"
    advice = "Maintain healthy lifestyle"
elif 120 <= systolic_bp <= 129 and diastolic_bp < 80:
    bp_category = "Elevated"
    advice = "Monitor regularly, lifestyle changes recommended"
elif 130 <= systolic_bp <= 139 or 80 <= diastolic_bp <= 89:
    bp_category = "Stage 1 Hypertension"
    advice = "Consult healthcare provider"
elif systolic_bp >= 140 or diastolic_bp >= 90:
    bp_category = "Stage 2 Hypertension"
    advice = "Medical treatment required"
else:
    bp_category = "Unknown"
    advice = "Please verify readings"

print(f"Classification: {bp_category}")
print(f"Recommendation: {advice}")
```

```python
# Grade Classification System
print("\n=== GRADING SYSTEM EXAMPLE ===")

# Test multiple scores
test_scores = [95, 82, 76, 68, 55, 45, 92, 88]

for score in test_scores:
    print(f"\nScore: {score}")
    
    if score >= 90:
        grade = 'A'
        performance = "Excellent"
    elif score >= 80:
        grade = 'B'
        performance = "Good"
    elif score >= 70:
        grade = 'C'
        performance = "Satisfactory"
    elif score >= 60:
        grade = 'D'
        performance = "Needs Improvement"
    else:
        grade = 'F'
        performance = "Unsatisfactory"
    
    print(f"  Grade: {grade}")
    print(f"  Performance: {performance}")
    
    # Additional feedback based on grade
    if grade in ['A', 'B']:
        print("  🎉 Keep up the great work!")
    elif grade == 'C':
        print("  📚 Room for improvement - review weak areas")
    else:
        print("  🎯 Additional support recommended")
```

```python
# Patient Risk Stratification
print("\n=== PATIENT RISK STRATIFICATION ===")

# Sample patient data
patients = [
    {'name': 'Alice', 'age': 45, 'smoker': False, 'bp': 125, 'cholesterol': 190},
    {'name': 'Bob', 'age': 62, 'smoker': True, 'bp': 145, 'cholesterol': 240},
    {'name': 'Carol', 'age': 35, 'smoker': False, 'bp': 118, 'cholesterol': 175},
    {'name': 'David', 'age': 58, 'smoker': False, 'bp': 135, 'cholesterol': 210}
]

for patient in patients:
    name = patient['name']
    age = patient['age']
    smoker = patient['smoker']
    bp = patient['bp']
    cholesterol = patient['cholesterol']
    
    print(f"\nPatient: {name}")
    print(f"  Age: {age}, Smoker: {smoker}, BP: {bp}, Cholesterol: {cholesterol}")
    
    # Calculate risk factors
    risk_factors = 0
    if age > 55:
        risk_factors += 1
    if smoker:
        risk_factors += 2
    if bp > 140:
        risk_factors += 1
    if cholesterol > 200:
        risk_factors += 1
    
    # Risk stratification
    if risk_factors == 0:
        risk_level = "Low"
        action = "Annual checkup sufficient"
    elif risk_factors <= 2:
        risk_level = "Moderate"
        action = "6-month follow-up recommended"
    elif risk_factors <= 4:
        risk_level = "High"
        action = "3-month monitoring required"
    else:
        risk_level = "Very High"
        action = "Immediate intervention needed"
    
    print(f"  Risk Factors: {risk_factors}")
    print(f"  Risk Level: {risk_level}")
    print(f"  Action: {action}")
```

## 4. Logical Operators - Combining Conditions

### What are Logical Operators?
**Logical operators** allow you to combine multiple conditions into complex expressions. They help you create more sophisticated decision logic.

### Key Operators:
- **`and`**: Both conditions must be `True`
- **`or`**: At least one condition must be `True`
- **`not`**: Inverts the condition (True becomes False, False becomes True)

### Truth Tables:

**AND Operator:**
| Condition 1 | Condition 2 | Result |
|-------------|-------------|---------|
| True | True | ✅ True |
| True | False | ❌ False |
| False | True | ❌ False |
| False | False | ❌ False |

**OR Operator:**
| Condition 1 | Condition 2 | Result |
|-------------|-------------|---------|
| True | True | ✅ True |
| True | False | ✅ True |
| False | True | ✅ True |
| False | False | ❌ False |

### Common Use Cases:
- Multi-criteria filtering
- Complex validation rules
- Compound risk factors
- Range checking

```python
# Logical Operators Examples
print("=== LOGICAL OPERATORS ===")

# Example 1: AND operator - All conditions must be true
print("\n1. AND Operator (All conditions must be true)")

patient_age = 35
has_symptoms = True
tested_positive = True

print(f"Age: {patient_age}, Has symptoms: {has_symptoms}, Tested positive: {tested_positive}")

if patient_age >= 18 and has_symptoms and tested_positive:
    print("✅ Patient qualifies for treatment protocol A")
    print("🩺 Begin immediate treatment")
else:
    print("❌ Patient does not meet all criteria")
    if patient_age < 18:
        print("👶 Use pediatric protocol instead")
    if not has_symptoms:
        print("😷 Monitor for symptom development")
    if not tested_positive:
        print("🧪 Consider retesting")
```

```python
# OR Operator Examples
print("\n2. OR Operator (At least one condition must be true)")

# Emergency criteria check
chest_pain = True
difficulty_breathing = False
severe_bleeding = False
loss_of_consciousness = False

print("Emergency Assessment:")
print(f"  Chest pain: {chest_pain}")
print(f"  Difficulty breathing: {difficulty_breathing}")
print(f"  Severe bleeding: {severe_bleeding}")
print(f"  Loss of consciousness: {loss_of_consciousness}")

if chest_pain or difficulty_breathing or severe_bleeding or loss_of_consciousness:
    print("\n🚨 EMERGENCY CONDITION DETECTED")
    print("➡️  Immediate medical attention required")
    print("📞 Call emergency services")
    
    # Determine specific emergency type
    if chest_pain:
        print("💔 Possible cardiac event")
    if difficulty_breathing:
        print("🫁 Possible respiratory emergency")
    if severe_bleeding:
        print("🩸 Hemorrhage control needed")
    if loss_of_consciousness:
        print("🧠 Neurological emergency")
else:
    print("\n✅ No emergency conditions detected")
    print("🩺 Proceed with standard assessment")
```

```python
# NOT Operator Examples
print("\n3. NOT Operator (Inverts the condition)")

# Data validation with NOT
patient_consent = False
insurance_verified = True

print(f"Patient consent: {patient_consent}")
print(f"Insurance verified: {insurance_verified}")

if not patient_consent:
    print("\n⚠️  Cannot proceed without patient consent")
    print("📝 Obtain signed consent form")
else:
    print("\n✅ Patient consent obtained")

if not insurance_verified:
    print("\n⚠️  Insurance not verified")
    print("💳 Payment arrangement needed")
else:
    print("\n✅ Insurance verified")

# Combined with other operators
if not patient_consent or not insurance_verified:
    print("\n⏸️  Procedure on hold - resolve issues above")
else:
    print("\n▶️  Ready to proceed with procedure")
```

```python
# Complex Logical Combinations
print("\n4. COMPLEX LOGICAL COMBINATIONS ===")

# Vaccine eligibility check
age = 45
is_essential_worker = True
has_underlying_condition = False
previous_dose_count = 2
months_since_last_dose = 8

print(f"Age: {age}")
print(f"Essential worker: {is_essential_worker}")
print(f"Underlying condition: {has_underlying_condition}")
print(f"Previous doses: {previous_dose_count}")
print(f"Months since last dose: {months_since_last_dose}")

# Complex eligibility logic
eligible_for_booster = False
priority_group = False

# Check if in priority group
if age >= 65 or has_underlying_condition or is_essential_worker:
    priority_group = True
    print("\n✅ Patient in priority group")

# Check booster eligibility
if previous_dose_count >= 2 and months_since_last_dose >= 6:
    eligible_for_booster = True
    print("✅ Minimum time since last dose met")

# Final decision
if priority_group and eligible_for_booster:
    print("\n🎯 RECOMMENDATION: Schedule booster shot")
    print("📅 Priority scheduling available")
elif eligible_for_booster and not priority_group:
    print("\n✅ RECOMMENDATION: Booster eligible")
    print("📅 Standard scheduling")
elif priority_group and not eligible_for_booster:
    months_to_wait = 6 - months_since_last_dose
    print(f"\n⏳ Wait {months_to_wait} more month(s) for booster")
else:
    print("\n❌ Not currently eligible for booster")
    print("📊 Continue monitoring guidelines")
```

## 5. Nested Conditionals and Decision Trees

### What are Nested Conditionals?
**Nested conditionals** are if statements inside other if statements. They allow for complex, multi-level decision making.

### Best Practices:
- ✅ **Limit Nesting**: Avoid going deeper than 2-3 levels
- ✅ **Clear Logic**: Each level should have a clear purpose
- ✅ **Comment**: Explain complex nested logic
- ✅ **Consider elif**: Sometimes elif is cleaner than nesting

### Common Use Cases:
- Multi-stage decision trees
- Hierarchical classification
- Progressive validation

```python
# Nested Conditional Examples
print("=== NESTED CONDITIONALS ===")

# Triage Decision Tree
patients_to_triage = [
    {'name': 'Patient A', 'conscious': True, 'breathing': True, 'severe_pain': True, 'bleeding': False},
    {'name': 'Patient B', 'conscious': False, 'breathing': True, 'severe_pain': False, 'bleeding': False},
    {'name': 'Patient C', 'conscious': True, 'breathing': True, 'severe_pain': False, 'bleeding': True},
    {'name': 'Patient D', 'conscious': True, 'breathing': False, 'severe_pain': False, 'bleeding': False}
]

for patient in patients_to_triage:
    name = patient['name']
    conscious = patient['conscious']
    breathing = patient['breathing']
    severe_pain = patient['severe_pain']
    bleeding = patient['bleeding']
    
    print(f"\n🏥 Triaging {name}:")
    
    # Level 1: Life-threatening conditions
    if not breathing or not conscious:
        priority = "CRITICAL"
        wait_time = "Immediate"
        
        if not breathing:
            action = "Initiate resuscitation protocol"
        else:
            action = "Neurological assessment, protect airway"
    
    # Level 2: Serious but not immediately life-threatening
    elif severe_pain or bleeding:
        priority = "HIGH"
        wait_time = "< 15 minutes"
        
        if severe_pain and bleeding:
            action = "Pain management + Hemorrhage control"
        elif severe_pain:
            action = "Immediate pain assessment and relief"
        else:
            action = "Wound assessment and treatment"
    
    # Level 3: Standard care
    else:
        priority = "STANDARD"
        wait_time = "30-60 minutes"
        action = "Routine assessment and treatment"
    
    print(f"  Priority: {priority}")
    print(f"  Wait time: {wait_time}")
    print(f"  Action: {action}")
```

```python
# Practical Data Processing with Conditionals
print("\n=== DATA PROCESSING WITH CONDITIONALS ===")

# Sample patient data with potential issues
raw_patient_data = [
    {'id': 1, 'name': 'Alice', 'age': 34, 'weight': 65, 'height': 165, 'bp': '120/80'},
    {'id': 2, 'name': 'Bob', 'age': -5, 'weight': 70, 'height': 175, 'bp': '140/90'},
    {'id': 3, 'name': 'Carol', 'age': 45, 'weight': 0, 'height': 160, 'bp': 'invalid'},
    {'id': 4, 'name': 'David', 'age': 28, 'weight': 80, 'height': 180, 'bp': '135/85'},
    {'id': 5, 'name': '', 'age': 52, 'weight': 75, 'height': 170, 'bp': '150/95'}
]

valid_records = []
invalid_records = []

for record in raw_patient_data:
    record_id = record['id']
    errors = []
    
    # Validate name
    if not record['name'] or record['name'].strip() == '':
        errors.append("Missing name")
    
    # Validate age
    if record['age'] < 0:
        errors.append(f"Invalid age: {record['age']} (negative)")
    elif record['age'] > 120:
        errors.append(f"Invalid age: {record['age']} (exceeds maximum)")
    
    # Validate weight
    if record['weight'] <= 0:
        errors.append(f"Invalid weight: {record['weight']} (must be positive)")
    elif record['weight'] > 500:
        errors.append(f"Invalid weight: {record['weight']} (exceeds maximum)")
    
    # Validate height
    if record['height'] <= 0:
        errors.append(f"Invalid height: {record['height']} (must be positive)")
    elif record['height'] > 250:
        errors.append(f"Invalid height: {record['height']} (exceeds maximum)")
    
    # Validate blood pressure format
    bp = record['bp']
    if not isinstance(bp, str) or '/' not in bp:
        errors.append(f"Invalid BP format: {bp}")
    else:
        try:
            systolic, diastolic = map(int, bp.split('/'))
            if not (50 <= systolic <= 250 and 30 <= diastolic <= 150):
                errors.append(f"BP values out of range: {bp}")
        except ValueError:
            errors.append(f"Invalid BP values: {bp}")
    
    # Classification
    if not errors:
        valid_records.append(record)
        print(f"✅ Record {record_id}: VALID")
    else:
        invalid_records.append({'record': record, 'errors': errors})
        print(f"❌ Record {record_id}: INVALID")
        for error in errors:
            print(f"   - {error}")

print(f"\n📊 Validation Summary:")
print(f"Valid records: {len(valid_records)}")
print(f"Invalid records: {len(invalid_records)}")
print(f"Success rate: {len(valid_records)/len(raw_patient_data)*100:.1f}%")
```

## 6. Best Practices and Common Pitfalls

### ✅ Best Practices:

1. **Use clear, descriptive conditions**
2. **Keep nesting shallow** (max 2-3 levels)
3. **Order conditions logically** (most specific first)
4. **Use parentheses** for complex conditions
5. **Test edge cases** thoroughly
6. **Comment complex logic**

### ❌ Common Pitfalls:

1. **Assignment vs Comparison**: Using `=` instead of `==`
2. **Indentation Errors**: Python relies on proper indentation
3. **Infinite Loops**: Conditions that never become False
4. **Overlapping Conditions**: Multiple conditions that could both be true
5. **Missing else**: Not handling all possible cases
6. **Too Deep Nesting**: Code becomes unreadable

```python
# Common Pitfall Demonstrations
print("=== COMMON PITFALLS AND SOLUTIONS ===")

# Pitfall 1: Assignment vs Comparison
print("\n1. Assignment (=) vs Comparison (==)")

value = 10

# WRONG: This assigns 5 to value and always evaluates to True
# if value = 5:  # SyntaxError in Python 3!
#     print("Value is 5")

# CORRECT: Use == for comparison
if value == 5:
    print("Value equals 5")
else:
    print(f"Value is {value}, not 5")

# Pitfall 2: Indentation Errors
print("\n2. Indentation Best Practices")

score = 85

# CORRECT: Consistent 4-space indentation
if score >= 90:
    grade = 'A'
    message = "Excellent"
elif score >= 80:
    grade = 'B'
    message = "Good"
else:
    grade = 'C'
    message = "Satisfactory"

print(f"Score: {score}, Grade: {grade}, Message: {message}")
```

```python
# Testing Edge Cases
print("\n3. Edge Case Testing")

# Boundary values
test_values = [0, 1, 17, 18, 64, 65, 100, 101]

for age in test_values:
    print(f"\nAge: {age}")
    
    # Test boundary conditions
    if age < 0:
        category = "Invalid"
    elif age < 13:
        category = "Child"
    elif age < 20:
        category = "Teenager"
    elif age < 65:
        category = "Adult"
    elif age <= 120:
        category = "Senior"
    else:
        category = "Invalid"
    
    print(f"  Category: {category}")
    
    # Additional checks
    if age == 18:
        print("  🎉 Congratulations! You are now an adult.")
    if age == 65:
        print("  🎂 Welcome to senior benefits eligibility.")
    if age < 0 or age > 120:
        print("  ⚠️  Please verify age value")
```

```python
# Final Verification: All Conditional Concepts
print("\n=== CONDITIONAL LOGIC VERIFICATION ===")

# Test all concepts work correctly
test_results = []

# Test 1: Basic if
if True:
    test_results.append(("Basic if", True))

# Test 2: if-else
if False:
    test_results.append(("if-else", False))
else:
    test_results.append(("if-else", True))

# Test 3: if-elif-else
value = 2
if value == 1:
    test_results.append(("if-elif-else", False))
elif value == 2:
    test_results.append(("if-elif-else", True))
else:
    test_results.append(("if-elif-else", False))

# Test 4: Logical operators
if True and True and not False:
    test_results.append(("Logical operators", True))

# Test 5: Comparison operators
if 5 > 3 and 3 < 5 and 5 >= 5 and 5 <= 5 and 5 == 5 and 5 != 3:
    test_results.append(("Comparison operators", True))

# Test 6: Nested conditionals
if True:
    if True:
        test_results.append(("Nested conditionals", True))

# Display results
passed = sum(1 for _, result in test_results if result)
total = len(test_results)

print(f"Verification Results: {passed}/{total} tests passed")
for test_name, result in test_results:
    status = "✅" if result else "❌"
    print(f"  {status} {test_name}")

if passed == total:
    print("\n🎉 All conditional logic concepts verified!")
else:
    print(f"\n⚠️  {total - passed} test(s) failed")
```
