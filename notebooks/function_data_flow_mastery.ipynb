## 1. Understanding Parameters and Arguments

### Key Distinction:
- **Parameters**: Variables in the function **definition** (the placeholders)
- **Arguments**: Actual values passed during function **calls** (the real data)

### Analogy:
Think of a function like a vending machine:
- **Parameters** = The coin slot and button (where inputs go)
- **Arguments** = The actual coins you insert and button you press
- **Return value** = The snack that comes out

### Why This Matters:
- ✅ **Flexibility**: Functions work with any valid input data
- ✅ **Reusability**: Same function, different data = different results
- ✅ **Testability**: Can test with various inputs
- ✅ **Modularity**: Functions are independent of specific data values

````python
# Parameters vs Arguments - Clear Demonstration
print("=== PARAMETERS vs ARGUMENTS ===")

# Function DEFINITION (parameters: weight, height)
def calculate_bmi(weight, height):
    """
    Calculate BMI from weight and height.
    
    PARAMETERS (in definition):
        weight: The parameter name (placeholder)
        height: The parameter name (placeholder)
    """
    bmi = weight / (height ** 2)
    return bmi

print("Function defined with PARAMETERS: weight, height")
print()

# Function CALLS (arguments: actual values)
print("Calling function with ARGUMENTS:")

# Call 1: Arguments are 70 and 1.75
bmi1 = calculate_bmi(70, 1.75)  # 70 → weight, 1.75 → height
print(f"  calculate_bmi(70, 1.75) → BMI = {bmi1:.2f}")

# Call 2: Different arguments
bmi2 = calculate_bmi(85, 1.80)  # 85 → weight, 1.80 → height
print(f"  calculate_bmi(85, 1.80) → BMI = {bmi2:.2f}")

# Call 3: Different arguments again
bmi3 = calculate_bmi(55, 1.65)  # 55 → weight, 1.65 → height
print(f"  calculate_bmi(55, 1.65) → BMI = {bmi3:.2f}")

print("\n✅ Same function (parameters), different data (arguments), different results!")
````

````python
# Multiple Parameters - Healthcare Examples
print("\n=== MULTIPLE PARAMETERS IN ACTION ===")

def assess_patient_risk(age, systolic_bp, smoker, family_history):
    """
    Assess cardiovascular risk based on multiple factors.
    
    PARAMETERS (placeholders in definition):
        age: Patient age in years
        systolic_bp: Systolic blood pressure
        smoker: True if patient smokes
        family_history: True if heart disease in family
    """
    risk_score = 0
    
    # Age factor
    if age > 65:
        risk_score += 3
    elif age > 45:
        risk_score += 2
    elif age > 30:
        risk_score += 1
    
    # BP factor
    if systolic_bp > 140:
        risk_score += 3
    elif systolic_bp > 130:
        risk_score += 2
    elif systolic_bp > 120:
        risk_score += 1
    
    # Lifestyle factors
    if smoker:
        risk_score += 3
    if family_history:
        risk_score += 2
    
    # Classify risk
    if risk_score >= 7:
        return "High Risk"
    elif risk_score >= 4:
        return "Moderate Risk"
    else:
        return "Low Risk"

print("Function defined with 4 parameters:")
print("  assess_patient_risk(age, systolic_bp, smoker, family_history)")
print()

# Different patients = different arguments
patients = [
    {"name": "Alice", "args": (34, 118, False, False)},
    {"name": "Bob", "args": (58, 145, True, True)},
    {"name": "Carol", "args": (45, 125, False, True)},
    {"name": "David", "args": (67, 150, True, False)},
]

print("Calling with different ARGUMENTS for each patient:")
for patient in patients:
    name = patient["name"]
    args = patient["args"]
    
    # Unpack arguments: age, bp, smoker, family = args
    risk = assess_patient_risk(*args)
    
    print(f"  {name} {args} → {risk}")

print("\n✅ One function, many different inputs, predictable outputs!")
````

## 2. Returning Values from Functions

### Why Return Values Matter:
- **🔧 Reusability**: Results can be used elsewhere in the program
- **🔗 Composability**: Output of one function becomes input to another
- **🧪 Testability**: Can verify function produces correct output
- **📦 Data Flow**: Information moves through the program via returns

### The return Statement:
- ✅ **Sends data back** to the caller
- ✅ **Ends function execution** immediately
- ✅ **Can return any data type** (numbers, strings, lists, dictionaries, objects)
- ✅ **Functions without return** implicitly return `None`

### ❌ Common Mistake: print() vs return()
```python
# BAD: Only prints, doesn't return
def bad_calculate_area(length, width):
    area = length * width
    print(f"Area: {area}")  # ❌ Can't use this value elsewhere!

# GOOD: Returns the value
def good_calculate_area(length, width):
    area = length * width
    return area  # ✅ Can store, use, or pass this value
```

````python
# Return vs Print - Critical Difference
print("=== RETURN vs PRINT - CRITICAL DIFFERENCE ===")

def calculate_area_print(length, width):
    """
    ❌ BAD: Only prints, doesn't return.
    The calculated value is LOST after printing!
    """
    area = length * width
    print(f"Calculated area: {area}")  # Output to screen only
    # No return statement!

def calculate_area_return(length, width):
    """
    ✅ GOOD: Returns the calculated value.
    The value can be captured, stored, and reused!
    """
    area = length * width
    return area  # Value is sent back to caller

print("Testing BAD version (print only):")
print("  Calling calculate_area_print(5, 3)...")
result_print = calculate_area_print(5, 3)
print(f"  Captured result: {result_print}")  # None!
print(f"  Type: {type(result_print)}")

print("\nTesting GOOD version (return):")
print("  Calling calculate_area_return(5, 3)...")
result_return = calculate_area_return(5, 3)
print(f"  Captured result: {result_return}")
print(f"  Type: {type(result_return)}")

print("\n=== WHY RETURN IS BETTER ===")
print("With returned value, we can:")

# 1. Store for later use
room_area = calculate_area_return(4, 5)
print(f"1. Store in variable: room_area = {room_area}")

# 2. Use in calculations
total_area = calculate_area_return(4, 5) + calculate_area_return(3, 4)
print(f"2. Use in calculation: {calculate_area_return(4, 5)} + {calculate_area_return(3, 4)} = {total_area}")

# 3. Pass to other functions
def calculate_paint_needed(area, coverage_per_liter=10):
    return area / coverage_per_liter

paint_needed = calculate_paint_needed(calculate_area_return(4, 5))
print(f"3. Pass to other function: Paint needed = {paint_needed:.2f} liters")

# 4. Use in conditions
if calculate_area_return(10, 10) > 50:
    print(f"4. Use in condition: Area {calculate_area_return(10, 10)} > 50 → Large room!")

print("\n✅ Return values make functions truly reusable!")
````

````python
# Multiple Return Values and Early Returns
print("\n=== ADVANCED RETURN PATTERNS ===")

# Returning multiple values as a tuple
def analyze_vitals(heart_rate, systolic, diastolic):
    """
    Analyze vital signs and return multiple assessments.
    
    Returns tuple: (hr_status, bp_status, overall_status)
    """
    # Heart rate assessment
    if 60 <= heart_rate <= 100:
        hr_status = "Normal"
    elif heart_rate < 60:
        hr_status = "Bradycardia"
    else:
        hr_status = "Tachycardia"
    
    # BP assessment
    if systolic < 120 and diastolic < 80:
        bp_status = "Normal"
    elif systolic < 140 or diastolic < 90:
        bp_status = "Elevated"
    else:
        bp_status = "Hypertensive"
    
    # Overall status
    if hr_status == "Normal" and bp_status == "Normal":
        overall = "Healthy"
    elif "Bradycardia" in hr_status or "Tachycardia" in hr_status or bp_status == "Hypertensive":
        overall = "Critical"
    else:
        overall = "Caution"
    
    # Return multiple values as tuple
    return hr_status, bp_status, overall

print("Function returning multiple values:")
hr_status, bp_status, overall = analyze_vitals(72, 118, 78)
print(f"  Heart Rate: {hr_status}, BP: {bp_status}, Overall: {overall}")

hr_status, bp_status, overall = analyze_vitals(55, 145, 92)
print(f"  Heart Rate: {hr_status}, BP: {bp_status}, Overall: {overall}")

# Returning dictionaries (more descriptive)
def comprehensive_vital_analysis(heart_rate, systolic, diastolic, temperature):
    """
    Comprehensive analysis returning structured data.
    """
    return {
        'heart_rate': {
            'value': heart_rate,
            'status': 'Normal' if 60 <= heart_rate <= 100 else 'Abnormal',
            'unit': 'bpm'
        },
        'blood_pressure': {
            'systolic': systolic,
            'diastolic': diastolic,
            'status': 'Normal' if systolic < 120 and diastolic < 80 else 'Elevated',
            'unit': 'mmHg'
        },
        'temperature': {
            'value': temperature,
            'status': 'Normal' if 97 <= temperature <= 99 else 'Fever' if temperature > 99 else 'Low',
            'unit': 'F'
        }
    }

print("\nFunction returning dictionary:")
analysis = comprehensive_vital_analysis(72, 118, 78, 98.6)
print(f"  Heart Rate: {analysis['heart_rate']['value']} {analysis['heart_rate']['unit']} ({analysis['heart_rate']['status']})")
print(f"  Blood Pressure: {analysis['blood_pressure']['systolic']}/{analysis['blood_pressure']['diastolic']} {analysis['blood_pressure']['unit']} ({analysis['blood_pressure']['status']})")
print(f"  Temperature: {analysis['temperature']['value']}°{analysis['temperature']['unit']} ({analysis['temperature']['status']})")

# Early return pattern (guard clauses)
def safe_divide(numerator, denominator):
    """
    Safely divide with early return for invalid cases.
    """
    # Guard clause - return early if denominator is zero
    if denominator == 0:
        return None  # Early return!
    
    # Main calculation only happens if guard passes
    return numerator / denominator

print("\nEarly return (guard clause) pattern:")
print(f"  safe_divide(10, 2) = {safe_divide(10, 2)}")
print(f"  safe_divide(10, 0) = {safe_divide(10, 0)}")
````

## 3. Using Returned Results - Building Data Pipelines

### The Power of Return Values:
Return values enable you to build **data pipelines** where the output of one function becomes the input to another. This is how complex programs are built from simple, reusable pieces.

### Data Flow Patterns:
```
Input → Function A → Return → Function B → Return → Function C → Output
```

### Benefits:
- ✅ **Composability**: Chain functions together
- ✅ **Testability**: Test each function independently
- ✅ **Flexibility**: Mix and match functions in different combinations
- ✅ **Clarity**: Each step is explicit and named

````python
# Data Pipeline Example: Patient Risk Assessment
print("=== DATA PIPELINE: PATIENT RISK ASSESSMENT ===")

# Step 1: Calculate BMI
def calculate_bmi(weight_kg, height_m):
    """Calculate and return BMI."""
    return weight_kg / (height_m ** 2)

# Step 2: Classify BMI
def classify_bmi(bmi):
    """Classify BMI value and return category."""
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Step 3: Calculate risk score from BMI category
def calculate_bmi_risk(bmi_category):
    """Return risk score based on BMI category."""
    risk_scores = {
        "Underweight": 1,
        "Normal": 0,
        "Overweight": 2,
        "Obese": 4
    }
    return risk_scores.get(bmi_category, 0)

# Step 4: Generate recommendation
def generate_recommendation(risk_score):
    """Generate health recommendation based on risk score."""
    if risk_score == 0:
        return "Maintain healthy lifestyle"
    elif risk_score <= 2:
        return "Consider dietary changes"
    else:
        return "Consult healthcare provider"

print("Building a data pipeline with 4 functions:")
print("  Step 1: calculate_bmi() → returns BMI value")
print("  Step 2: classify_bmi() → returns BMI category")
print("  Step 3: calculate_bmi_risk() → returns risk score")
print("  Step 4: generate_recommendation() → returns advice")
print()

# Patient data
patients = [
    {"name": "Alice", "weight": 65, "height": 1.70},
    {"name": "Bob", "weight": 85, "height": 1.75},
    {"name": "Carol", "weight": 55, "height": 1.65},
]

print("Processing patients through the pipeline:")
for patient in patients:
    name = patient["name"]
    weight = patient["weight"]
    height = patient["height"]
    
    # PIPELINE: Output of each function becomes input to next
    bmi = calculate_bmi(weight, height)                    # Step 1
    bmi_category = classify_bmi(bmi)                        # Step 2 (uses bmi)
    risk_score = calculate_bmi_risk(bmi_category)           # Step 3 (uses category)
    recommendation = generate_recommendation(risk_score)    # Step 4 (uses score)
    
    print(f"\n  {name}:")
    print(f"    BMI: {bmi:.1f} → Category: {bmi_category}")
    print(f"    Risk Score: {risk_score}")
    print(f"    Recommendation: {recommendation}")

print("\n=== CHAINING FUNCTIONS (One-liner) ===")
print("Same pipeline as a chain:")

for patient in patients:
    # Chained version - less readable but shows data flow
    final_rec = generate_recommendation(
        calculate_bmi_risk(
            classify_bmi(
                calculate_bmi(patient["weight"], patient["height"])
            )
        )
    )
    print(f"  {patient['name']}: {final_rec}")

print("\n✅ Each function does one thing, return values connect them!")
````

````python
# Complex Data Pipeline: Patient Processing System
print("\n=== COMPLEX DATA PIPELINE: PATIENT PROCESSING ===")

# Stage 1: Data Collection
def collect_patient_data():
    """Simulate collecting patient data from various sources."""
    return [
        {"id": "P001", "name": "John Doe", "age": 45, "weight": 80, "height": 1.75, "bp": "135/85", "hr": 72},
        {"id": "P002", "name": "Jane Smith", "age": 34, "weight": 65, "height": 1.70, "bp": "120/80", "hr": 68},
        {"id": "P003", "name": "Bob Johnson", "age": 67, "weight": 90, "height": 1.80, "bp": "145/90", "hr": 78},
    ]

# Stage 2: Data Validation
def validate_patient_record(record):
    """Validate a patient record and return (is_valid, errors)."""
    errors = []
    
    if not record.get("name"):
        errors.append("Missing name")
    
    age = record.get("age", 0)
    if not (0 <= age <= 120):
        errors.append(f"Invalid age: {age}")
    
    return len(errors) == 0, errors

# Stage 3: Data Enrichment
def enrich_patient_data(record):
    """Add calculated fields to patient record."""
    enriched = record.copy()
    
    # Calculate BMI
    weight = record["weight"]
    height = record["height"]
    enriched["bmi"] = round(weight / (height ** 2), 1)
    
    # Parse BP
    systolic, diastolic = map(int, record["bp"].split("/"))
    enriched["systolic"] = systolic
    enriched["diastolic"] = diastolic
    
    # Calculate age group
    age = record["age"]
    if age < 18:
        enriched["age_group"] = "Minor"
    elif age < 65:
        enriched["age_group"] = "Adult"
    else:
        enriched["age_group"] = "Senior"
    
    return enriched

# Stage 4: Risk Assessment
def assess_health_risks(enriched_record):
    """Assess health risks and return risk assessment."""
    risks = []
    risk_score = 0
    
    # BMI risk
    bmi = enriched_record["bmi"]
    if bmi > 30:
        risks.append("Obesity")
        risk_score += 3
    elif bmi > 25:
        risks.append("Overweight")
        risk_score += 1
    
    # BP risk
    systolic = enriched_record["systolic"]
    if systolic > 140:
        risks.append("Hypertension")
        risk_score += 3
    elif systolic > 130:
        risks.append("Elevated BP")
        risk_score += 1
    
    # Age risk
    if enriched_record["age"] > 65:
        risks.append("Advanced age")
        risk_score += 1
    
    return {
        "risk_score": risk_score,
        "risk_factors": risks,
        "risk_level": "High" if risk_score >= 5 else "Moderate" if risk_score >= 2 else "Low"
    }

# Stage 5: Report Generation
def generate_patient_report(enriched_record, risk_assessment):
    """Generate final patient report."""
    return {
        "patient_id": enriched_record["id"],
        "name": enriched_record["name"],
        "summary": {
            "age": enriched_record["age"],
            "bmi": enriched_record["bmi"],
            "bp": enriched_record["bp"],
            "age_group": enriched_record["age_group"]
        },
        "risk_assessment": risk_assessment,
        "recommendations": generate_recommendations_from_risk(risk_assessment)
    }

def generate_recommendations_from_risk(risk_assessment):
    """Generate recommendations based on risk."""
    level = risk_assessment["risk_level"]
    factors = risk_assessment["risk_factors"]
    
    recommendations = []
    
    if level == "High":
        recommendations.append("Schedule immediate follow-up")
        recommendations.append("Consider specialist referral")
    elif level == "Moderate":
        recommendations.append("Schedule routine follow-up")
        recommendations.append("Monitor closely")
    else:
        recommendations.append("Continue regular checkups")
    
    for factor in factors:
        if factor == "Obesity" or factor == "Overweight":
            recommendations.append("Consider weight management program")
        if factor == "Hypertension":
            recommendations.append("Blood pressure monitoring and lifestyle changes")
    
    return recommendations

# Run the complete pipeline
print("Running 5-stage patient processing pipeline...")
print()

# Stage 1: Collect
raw_data = collect_patient_data()
print(f"Stage 1 - Collected {len(raw_data)} patient records")

# Stage 2: Validate
valid_patients = []
for record in raw_data:
    is_valid, errors = validate_patient_record(record)
    if is_valid:
        valid_patients.append(record)
    else:
        print(f"  ⚠️  Invalid record {record.get('id', 'unknown')}: {errors}")

print(f"Stage 2 - Validated: {len(valid_patients)}/{len(raw_data)} records valid")

# Stage 3: Enrich
enriched_patients = [enrich_patient_data(p) for p in valid_patients]
print(f"Stage 3 - Enriched {len(enriched_patients)} records with calculated fields")

# Stage 4: Assess Risks
risk_assessments = [assess_health_risks(p) for p in enriched_patients]
print(f"Stage 4 - Assessed risks for {len(risk_assessments)} patients")

# Stage 5: Generate Reports
final_reports = [
    generate_patient_report(enriched_patients[i], risk_assessments[i])
    for i in range(len(enriched_patients))
]
print(f"Stage 5 - Generated {len(final_reports)} final reports")

# Display results
print("\n=== FINAL REPORTS ===")
for report in final_reports:
    print(f"\n{report['name']} ({report['patient_id']}):")
    print(f"  Age: {report['summary']['age']} ({report['summary']['age_group']})")
    print(f"  BMI: {report['summary']['bmi']}")
    print(f"  BP: {report['summary']['bp']}")
    print(f"  Risk Level: {report['risk_assessment']['risk_level']}")
    print(f"  Risk Factors: {', '.join(report['risk_assessment']['risk_factors']) or 'None'}")
    print(f"  Recommendations:")
    for rec in report['recommendations']:
        print(f"    • {rec}")

print("\n✅ Complex data pipeline completed successfully!")
````

## 4. Common Mistakes and How to Avoid Them

### Mistake 1: Hardcoding Values Inside Functions
Functions should be flexible and work with any valid input, not just specific hardcoded values.

### Mistake 2: Mixing print and return Incorrectly
Using print() when you should return, or vice versa.

### Mistake 3: Missing Return Statements
Forgetting to return a value, causing the function to return None unexpectedly.

### Mistake 4: Not Using Returned Values
Calling a function that returns a value but not capturing or using the result.

### Mistake 5: Inconsistent Return Types
Returning different types from different code paths (e.g., number in one path, None in another).

````python
# Common Mistakes Demonstrations
print("=== COMMON MISTAKES AND SOLUTIONS ===")

print("\n❌ MISTAKE 1: Hardcoding Values")
print("  BAD - Function only works for one specific patient:")

def bad_calculate_patient_bmi():
    weight = 70  # ❌ Hardcoded!
    height = 1.75  # ❌ Hardcoded!
    return weight / (height ** 2)

bmi1 = bad_calculate_patient_bmi()
print(f"    Result: {bmi1:.2f}")
print("    Problem: Can't calculate BMI for any other patient!")

print("\n  GOOD - Function accepts any weight and height:")

def good_calculate_bmi(weight, height):
    return weight / (height ** 2)

bmi2 = good_calculate_bmi(70, 1.75)  # Patient 1
bmi3 = good_calculate_bmi(85, 1.80)  # Patient 2
bmi4 = good_calculate_bmi(55, 1.60)  # Patient 3
print(f"    Patient 1: {bmi2:.2f}")
print(f"    Patient 2: {bmi3:.2f}")
print(f"    Patient 3: {bmi4:.2f}")
print("    ✓ Works for ANY patient with valid data!")

print("\n❌ MISTAKE 2: Print Instead of Return")
print("  BAD - Function only prints, can't use the value:")

def bad_double_value(x):
    result = x * 2
    print(f"Result: {result}")  # ❌ Value is lost after printing

output = bad_double_value(5)
print(f"    Captured: {output}")  # None!
# Can't do: final = bad_double_value(5) + 10  # Would crash!

print("\n  GOOD - Function returns the value:")

def good_double_value(x):
    return x * 2  # ✅ Value is sent back to caller

returned_value = good_double_value(5)
print(f"    Captured: {returned_value}")
final = good_double_value(5) + 10  # ✅ Works!
print(f"    Can use in calculation: {returned_value} + 10 = {final}")
````

````python
print("\n❌ MISTAKE 3: Missing Return Statements")
print("  BAD - Forgetting to return in some code paths:")

def bad_find_max(a, b):
    if a > b:
        return a  # Returns if a > b
    # ❌ What if a <= b? No return! Returns None implicitly!

result1 = bad_find_max(10, 5)
result2 = bad_find_max(3, 8)
print(f"    max(10, 5) = {result1} ✓")
print(f"    max(3, 8) = {result2} ✗ (None!)")

print("\n  GOOD - Always return a value:")

def good_find_max(a, b):
    if a > b:
        return a
    else:
        return b  # ✅ Always returns something

result3 = good_find_max(10, 5)
result4 = good_find_max(3, 8)
print(f"    max(10, 5) = {result3} ✓")
print(f"    max(3, 8) = {result4} ✓")

print("\n❌ MISTAKE 4: Ignoring Returned Values")
print("  BAD - Calling function but not using result:")

def get_important_data():
    return "Critical Patient Alert!"

get_important_data()  # ❌ Called but result is lost!
print("    Function called but result not captured or used!")

print("\n  GOOD - Capture and use the returned value:")

alert = get_important_data()  # ✅ Captured!
print(f"    Captured: {alert}")
print(f"    Can use it: ALERT: {alert.upper()}")

print("\n❌ MISTAKE 5: Inconsistent Return Types")
print("  BAD - Returning different types from different paths:")

def bad_calculate_discount(price, is_member):
    if is_member:
        return price * 0.9  # Returns float
    # ❌ No explicit return = returns None!

member_price = bad_calculate_discount(100, True)
non_member_price = bad_calculate_discount(100, False)
print(f"    Member: {member_price} (type: {type(member_price).__name__})")
print(f"    Non-member: {non_member_price} (type: {type(non_member_price).__name__}) ✗")

print("\n  GOOD - Consistent return type:")

def good_calculate_discount(price, is_member):
    if is_member:
        return price * 0.9
    return price  # ✅ Always returns a number

member_price2 = good_calculate_discount(100, True)
non_member_price2 = good_calculate_discount(100, False)
print(f"    Member: {member_price2} (type: {type(member_price2).__name__})")
print(f"    Non-member: {non_member_price2} (type: {type(non_member_price2).__name__}) ✓")
````

## 5. Best Practices for Function Data Flow

### ✅ Do's:

1. **Use Parameters, Not Hardcoding**
   - Make functions work with any valid input
   - Pass data as arguments, don't embed in function

2. **Return Values, Don't Just Print**
   - Use `return` to send data back
   - Use `print` only for user-facing output

3. **Capture and Use Returned Values**
   - Store return values in variables
   - Pass returned values to other functions
   - Use in calculations and conditions

4. **Keep Return Types Consistent**
   - All code paths should return same type
   - Use None intentionally, not by accident

5. **Design for Composability**
   - Output of function A should work as input to function B
   - Keep functions focused on single tasks

6. **Use Clear Parameter Names**
   - Names should describe what data is expected
   - `patient_age` is better than `a`

7. **Document Input and Output**
   - Docstrings should explain parameters and return values
   - Helps other developers (and future you!)

````python
# Best Practices Summary with Examples
print("=== BEST PRACTICES DEMONSTRATED ===")

# 1. Clear parameter names
def assess_cardiovascular_risk(patient_age, systolic_blood_pressure, is_smoker, has_family_history):
    """
    Assess cardiovascular risk based on patient factors.
    
    Args:
        patient_age: Age in years (int or float)
        systolic_blood_pressure: Systolic BP in mmHg (int)
        is_smoker: True if patient smokes (bool)
        has_family_history: True if family has heart disease (bool)
    
    Returns:
        dict: {
            'risk_score': int (0-10),
            'risk_level': str ('Low', 'Moderate', 'High'),
            'recommendation': str
        }
    """
    score = 0
    factors = []
    
    # Age scoring
    if patient_age >= 65:
        score += 3
        factors.append('Advanced age')
    elif patient_age >= 45:
        score += 2
    elif patient_age >= 30:
        score += 1
    
    # BP scoring
    if systolic_blood_pressure >= 140:
        score += 3
        factors.append('High blood pressure')
    elif systolic_blood_pressure >= 130:
        score += 2
        factors.append('Elevated blood pressure')
    
    # Lifestyle
    if is_smoker:
        score += 3
        factors.append('Smoking')
    
    if has_family_history:
        score += 2
        factors.append('Family history')
    
    # Determine level
    if score >= 7:
        level = "High"
        rec = "Immediate consultation recommended"
    elif score >= 4:
        level = "Moderate"
        rec = "Schedule follow-up within 3 months"
    else:
        level = "Low"
        rec = "Continue annual checkups"
    
    # Always return consistent structure
    return {
        'risk_score': score,
        'risk_level': level,
        'risk_factors': factors,
        'recommendation': rec
    }

# 2. Use returned values in chain
print("1. Using returned values in a chain:")

def get_patient_weight(patient_id):
    """Simulate getting patient weight."""
    weights = {'P001': 70, 'P002': 85, 'P003': 65}
    return weights.get(patient_id, 0)

def get_patient_height(patient_id):
    """Simulate getting patient height."""
    heights = {'P001': 1.75, 'P002': 1.80, 'P003': 1.70}
    return heights.get(patient_id, 0)

def calculate_bmi_from_patient_id(patient_id):
    """Calculate BMI using other functions' return values."""
    weight = get_patient_weight(patient_id)  # Use returned value
    height = get_patient_height(patient_id)  # Use returned value
    return weight / (height ** 2) if height > 0 else 0

for pid in ['P001', 'P002', 'P003']:
    bmi = calculate_bmi_from_patient_id(pid)
    print(f"    {pid}: BMI = {bmi:.2f}")

print("\n2. Consistent return structure example:")

# Good: All return paths return the same type
def safe_divide_all_paths(numerator, denominator):
    """Always returns a dictionary with success status."""
    if denominator == 0:
        return {'success': False, 'result': None, 'error': 'Division by zero'}
    
    if not isinstance(numerator, (int, float)) or not isinstance(denominator, (int, float)):
        return {'success': False, 'result': None, 'error': 'Invalid input type'}
    
    return {'success': True, 'result': numerator / denominator, 'error': None}

print("    Division with consistent return structure:")
for test in [(10, 2), (10, 0), ('a', 2)]:
    result = safe_divide_all_paths(*test)
    print(f"      {test[0]} / {test[1]}: success={result['success']}, result={result['result']}")

print("\n✅ Following best practices makes functions reliable and composable!")
````
