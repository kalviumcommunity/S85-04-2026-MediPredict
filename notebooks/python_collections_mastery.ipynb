## 1. Python Lists - Ordered and Mutable Collections

### What is a List?
A **list** is an ordered, mutable collection that can store multiple items. Lists are one of the most versatile data structures in Python.

### Key Characteristics:
- ✅ **Ordered**: Items maintain their position
- ✅ **Mutable**: Can be modified after creation
- ✅ **Indexed**: Access items by position (0-based)
- ✅ **Heterogeneous**: Can store different data types
- ✅ **Dynamic**: Can grow or shrink as needed

### Common Use Cases:
- Storing sequences of related data
- Collecting results from operations
- Managing dynamic collections that change over time

````python
# Creating Lists
print("=== CREATING LISTS ===")

# Empty list
empty_list = []
print(f"Empty list: {empty_list}")

# List with integers
patient_ages = [45, 32, 67, 28, 54, 41]
print(f"Patient ages: {patient_ages}")

# List with strings
diagnoses = ['Diabetes', 'Hypertension', 'Asthma', 'Flu', 'COVID-19']
print(f"Diagnoses: {diagnoses}")

# List with mixed data types
patient_record = ['John Doe', 45, 'Diabetes', 72.5, True]
print(f"Patient record: {patient_record}")

# List created with list() function
vitals = list([120, 80, 98.6, 72])
print(f"Vitals: {vitals}")
````

````python
# Accessing List Elements
print("\n=== ACCESSING LIST ELEMENTS ===")

blood_pressures = [120, 135, 142, 118, 128, 145, 132]
print(f"Blood pressures: {blood_pressures}")

# Indexing (0-based)
print(f"First reading: {blood_pressures[0]}")
print(f"Second reading: {blood_pressures[1]}")
print(f"Last reading: {blood_pressures[-1]}")
print(f"Second to last: {blood_pressures[-2]}")

# Slicing
print(f"First 3 readings: {blood_pressures[0:3]}")
print(f"Readings 2-5: {blood_pressures[1:5]}")
print(f"Every other reading: {blood_pressures[::2]}")
print(f"Reversed readings: {blood_pressures[::-1]}")
````

````python
# Modifying Lists
print("\n=== MODIFYING LISTS (MUTABILITY) ===")

medications = ['Aspirin', 'Metformin', 'Lisinopril']
print(f"Original medications: {medications}")

# Changing an element
medications[1] = 'Insulin'
print(f"After modification: {medications}")

# Adding elements
medications.append('Atorvastatin')
print(f"After append: {medications}")

medications.insert(1, 'Omeprazole')
print(f"After insert: {medications}")

# Extending with another list
new_meds = ['Vitamin D', 'Calcium']
medications.extend(new_meds)
print(f"After extend: {medications}")

# Removing elements
medications.remove('Aspirin')
print(f"After remove: {medications}")

removed_med = medications.pop()
print(f"Popped: {removed_med}, Remaining: {medications}")

# Clearing the list
medications.clear()
print(f"After clear: {medications}")
````

````python
# List Operations and Methods
print("\n=== LIST OPERATIONS ===")

patient_scores = [85, 92, 78, 95, 88, 91, 73, 89]
print(f"Patient scores: {patient_scores}")

# Length
print(f"Number of patients: {len(patient_scores)}")

# Sorting
sorted_scores = sorted(patient_scores)
print(f"Sorted scores: {sorted_scores}")

# Reverse sort in place
patient_scores.sort(reverse=True)
print(f"Sorted descending: {patient_scores}")

# Min, Max, Sum
print(f"Lowest score: {min(patient_scores)}")
print(f"Highest score: {max(patient_scores)}")
print(f"Average score: {sum(patient_scores) / len(patient_scores):.2f}")

# Checking membership
print(f"Is 95 in scores? {95 in patient_scores}")
print(f"Count of 88: {patient_scores.count(88)}")
print(f"Index of 95: {patient_scores.index(95)}")
````

## 2. Python Tuples - Ordered and Immutable Collections

### What is a Tuple?
A **tuple** is an ordered, immutable collection. Once created, a tuple cannot be modified.

### Key Characteristics:
- ✅ **Ordered**: Items maintain their position
- ✅ **Immutable**: Cannot be modified after creation
- ✅ **Indexed**: Access items by position (0-based)
- ✅ **Hashable**: Can be used as dictionary keys
- ✅ **Memory Efficient**: More compact than lists

### Common Use Cases:
- Storing fixed data that should not change
- Function return values with multiple items
- Dictionary keys (when you need composite keys)
- Data integrity protection

````python
# Creating Tuples
print("=== CREATING TUPLES ===")

# Empty tuple
empty_tuple = ()
print(f"Empty tuple: {empty_tuple}")

# Tuple with integers
vital_signs = (120, 80, 98.6, 72)
print(f"Vital signs: {vital_signs}")

# Tuple with strings
patient_info = ('John', 'Doe', 'Male', 'A+')
print(f"Patient info: {patient_info}")

# Single element tuple (note the comma!)
single_value = (42,)
print(f"Single element tuple: {single_value}")
print(f"Type: {type(single_value)}")

# Without comma, it's not a tuple
not_a_tuple = (42)
print(f"Without comma: {not_a_tuple}, Type: {type(not_a_tuple)}")

# Tuple from list
medications_list = ['Aspirin', 'Metformin']
medications_tuple = tuple(medications_list)
print(f"Tuple from list: {medications_tuple}")
````

````python
# Accessing Tuple Elements
print("\n=== ACCESSING TUPLE ELEMENTS ===")

lab_results = (7.2, 140, 95, 210, 1.2)
print(f"Lab results: {lab_results}")

# Indexing (same as lists)
print(f"Hemoglobin: {lab_results[0]}")
print(f"Cholesterol: {lab_results[3]}")
print(f"Last value: {lab_results[-1]}")

# Slicing (same as lists)
print(f"First 3 values: {lab_results[:3]}")
print(f"Last 2 values: {lab_results[-2:]}")

# Unpacking tuples
systolic, diastolic, temp, hr = 120, 80, 98.6, 72
print(f"\nUnpacked vitals: BP {systolic}/{diastolic}, Temp {temp}, HR {hr}")

# Tuple unpacking with *
first, *middle, last = (1, 2, 3, 4, 5)
print(f"First: {first}, Middle: {middle}, Last: {last}")
````

````python
# Demonstrating Immutability
print("\n=== DEMONSTRATING IMMUTABILITY ===")

medical_record = ('John Doe', 45, 'Diabetes')
print(f"Original medical record: {medical_record}")
print(f"Type: {type(medical_record)}")

# Try to modify (this will fail)
print("\nAttempting to modify tuple...")
try:
    medical_record[1] = 46
    print("Modified successfully (unexpected!)")
except TypeError as e:
    print(f"❌ Error (expected): {e}")
    print("✅ Tuples are immutable - cannot be modified!")

# However, we can create a new tuple
updated_record = ('John Doe', 46, 'Diabetes')
print(f"\nNew tuple created: {updated_record}")
print(f"Original unchanged: {medical_record}")

# Tuples with mutable objects
complex_tuple = ([1, 2, 3], 'immutable string')
print(f"\nComplex tuple: {complex_tuple}")
complex_tuple[0].append(4)  # This works!
print(f"After modifying list inside tuple: {complex_tuple}")
print("Note: The tuple is immutable, but mutable objects inside can still change!")
````

````python
# Tuple Operations and Methods
print("\n=== TUPLE OPERATIONS ===")

dimensions = (1920, 1080, 60)  # width, height, fps
print(f"Video dimensions: {dimensions}")

# Length
print(f"Number of dimensions: {len(dimensions)}")

# Count and Index
repeated_values = (1, 2, 2, 3, 2, 4)
print(f"\nRepeated values: {repeated_values}")
print(f"Count of 2: {repeated_values.count(2)}")
print(f"Index of first 3: {repeated_values.index(3)}")

# Concatenation (creates new tuple)
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined = tuple1 + tuple2
print(f"\nTuple 1: {tuple1}")
print(f"Tuple 2: {tuple2}")
print(f"Combined: {combined}")
print(f"Original tuples unchanged: {tuple1}, {tuple2}")

# Repetition
repeated = ('x',) * 5
print(f"\nRepeated tuple: {repeated}")

# Membership testing
print(f"Is 3 in combined? {3 in combined}")
print(f"Is 10 in combined? {10 in combined}")
````

## 3. Python Dictionaries - Key-Value Pair Collections

### What is a Dictionary?
A **dictionary** is an unordered (insertion-ordered in Python 3.7+), mutable collection of key-value pairs. Each key must be unique.

### Key Characteristics:
- ✅ **Key-Value Pairs**: Each item has a key and a value
- ✅ **Mutable**: Can be modified after creation
- ✅ **Key-Based Access**: Access values by key, not index
- ✅ **Unique Keys**: Each key must be unique
- ✅ **Hashable Keys**: Keys must be immutable (strings, numbers, tuples)
- ✅ **Dynamic**: Can grow or shrink as needed

### Common Use Cases:
- Storing structured data about objects
- Fast lookups by key
- Counting and grouping operations
- Configuration and settings storage
- Representing real-world entities

````python
# Creating Dictionaries
print("=== CREATING DICTIONARIES ===")

# Empty dictionary
empty_dict = {}
empty_dict2 = dict()
print(f"Empty dict: {empty_dict}")

# Dictionary with string keys
patient = {
    'name': 'Jane Smith',
    'age': 34,
    'gender': 'Female',
    'blood_type': 'O+',
    'allergies': ['Penicillin', 'Latex']
}
print(f"\nPatient dictionary:")
for key, value in patient.items():
    print(f"  {key}: {value}")

# Dictionary with different key types
mixed_keys = {
    'string_key': 'value1',
    42: 'integer key',
    (1, 2): 'tuple key'  # Tuples can be keys (immutable)
}
print(f"\nMixed key types: {mixed_keys}")

# Dictionary from lists using zip
fields = ['id', 'name', 'age']
values = [101, 'Bob Wilson', 45]
patient_from_zip = dict(zip(fields, values))
print(f"\nDictionary from zip: {patient_from_zip}")

# Dictionary comprehension
squares = {x: x**2 for x in range(1, 6)}
print(f"\nSquares dict: {squares}")
````

````python
# Accessing Dictionary Values
print("\n=== ACCESSING DICTIONARY VALUES ===")

vitals = {
    'systolic': 120,
    'diastolic': 80,
    'temperature': 98.6,
    'heart_rate': 72,
    'oxygen_sat': 98
}
print(f"Vitals dictionary: {vitals}")

# Access by key using square brackets
print(f"\nSystolic BP: {vitals['systolic']}")
print(f"Heart rate: {vitals['heart_rate']}")

# Using get() method (safer - no error if key missing)
print(f"Temperature: {vitals.get('temperature')}")
print(f"Respiratory rate: {vitals.get('respiratory_rate', 'Not recorded')}")

# Try to access non-existent key (will fail)
print("\nAttempting to access non-existent key...")
try:
    print(vitals['weight'])
except KeyError as e:
    print(f"❌ KeyError: {e}")
    print("✅ Key doesn't exist - use get() to handle missing keys safely!")
````

````python
# Modifying Dictionaries
print("\n=== MODIFYING DICTIONARIES ===")

medication = {
    'name': 'Metformin',
    'dosage': '500mg',
    'frequency': 'twice daily'
}
print(f"Original medication: {medication}")

# Adding new key-value pair
medication['route'] = 'oral'
print(f"After adding route: {medication}")

# Modifying existing value
medication['dosage'] = '1000mg'
print(f"After updating dosage: {medication}")

# Using update() to add multiple items
additional_info = {'start_date': '2024-01-15', 'prescribing_doctor': 'Dr. Johnson'}
medication.update(additional_info)
print(f"After update: {medication}")

# Removing items
removed_route = medication.pop('route')
print(f"Removed route: {removed_route}")
print(f"After pop: {medication}")

# Remove arbitrary item (Python 3.7+ removes last)
last_item = medication.popitem()
print(f"Removed last item: {last_item}")
print(f"After popitem: {medication}")

# Delete specific key
del medication['frequency']
print(f"After del: {medication}")

# Clear all items
medication.clear()
print(f"After clear: {medication}")
````

````python
# Dictionary Operations and Methods
print("\n=== DICTIONARY OPERATIONS ===")

lab_report = {
    'hemoglobin': 13.5,
    'glucose': 95,
    'cholesterol': 200,
    'hdl': 55,
    'ldl': 130,
    'triglycerides': 150
}

print(f"Lab report: {lab_report}")
print(f"\nNumber of tests: {len(lab_report)}")

# Keys, Values, Items
print(f"\nKeys: {list(lab_report.keys())}")
print(f"Values: {list(lab_report.values())}")
print(f"Items: {list(lab_report.items())[:3]}...")  # First 3 items

# Iterating over dictionary
print("\nIterating over key-value pairs:")
for test, value in lab_report.items():
    print(f"  {test}: {value}")

# Membership testing (checks keys)
print(f"\nIs 'glucose' a key? {'glucose' in lab_report}")
print(f"Is 'creatinine' a key? {'creatinine' in lab_report}")
print(f"Is 95 a value? {95 in lab_report.values()}")

# Nested dictionaries
patients_db = {
    'patient_001': {'name': 'Alice', 'age': 30, 'condition': 'Healthy'},
    'patient_002': {'name': 'Bob', 'age': 45, 'condition': 'Diabetes'},
    'patient_003': {'name': 'Carol', 'age': 55, 'condition': 'Hypertension'}
}

print(f"\nPatients database:")
for patient_id, info in patients_db.items():
    print(f"  {patient_id}: {info['name']}, {info['age']} years, {info['condition']}")
````

## 4. Choosing the Right Data Structure

### When to Use Each Collection

#### 📝 Lists - Use When:
- Data is ordered and sequential
- You need to modify elements frequently
- Order matters and you access by position
- You need to add/remove items dynamically
- Example: Patient visit history, test scores sequence

#### 🔒 Tuples - Use When:
- Data should not change (fixed values)
- You need data integrity protection
- You want to use data as dictionary keys
- Memory efficiency is important
- Example: Patient ID, coordinates, configuration settings

#### 📖 Dictionaries - Use When:
- You have labeled data with key-value relationships
- You need fast lookup by name/key
- Data represents attributes of an object
- You need to model structured entities
- Example: Patient records, configuration, lab results by test name

````python
# Practical Examples: Choosing the Right Structure
print("=== PRACTICAL EXAMPLES: CHOOSING RIGHT STRUCTURE ===")

# Example 1: List - Ordered collection that changes
print("\n1. LIST - Patient Visit Timeline:")
visit_dates = ['2024-01-15', '2024-02-20', '2024-03-18', '2024-04-22']
print(f"Visit dates: {visit_dates}")
visit_dates.append('2024-05-30')  # Add new visit
print(f"After adding new visit: {visit_dates}")
print("Use LIST when order matters and collection grows/changes")

# Example 2: Tuple - Fixed values that shouldn't change
print("\n2. TUPLE - Patient Demographics (fixed):")
patient_demographics = ('John Doe', '1990-05-15', 'M', 'A+')
name, dob, gender, blood_type = patient_demographics
print(f"Patient: {name}, DOB: {dob}, Gender: {gender}, Blood: {blood_type}")
print("Use TUPLE when data is fixed and shouldn't be modified")

# Example 3: Dictionary - Structured data with labels
print("\n3. DICTIONARY - Complete Patient Record:")
patient_record = {
    'patient_id': 'P001234',
    'name': 'Jane Smith',
    'age': 34,
    'vitals': {'bp': '120/80', 'hr': 72, 'temp': 98.6},
    'medications': ['Lisinopril', 'Metformin'],
    'allergies': ['Penicillin'],
    'emergency_contact': {'name': 'Bob Smith', 'phone': '555-0123'}
}

print(f"Patient: {patient_record['name']}")
print(f"Vitals: {patient_record['vitals']}")
print(f"Emergency contact: {patient_record['emergency_contact']['name']}")
print("Use DICTIONARY when you have structured data with named attributes")
````

````python
# Comparison Table: List vs Tuple vs Dictionary
print("\n=== COMPARISON: LIST vs TUPLE vs DICTIONARY ===")

comparison_data = [
    ['Characteristic', 'List []', 'Tuple ()', 'Dict {}'],
    ['Ordered', '✅ Yes', '✅ Yes', '✅ Yes (3.7+)'],
    ['Mutable', '✅ Yes', '❌ No', '✅ Yes'],
    ['Indexed', '✅ By position', '✅ By position', '❌ By key'],
    ['Duplicate items', '✅ Allowed', '✅ Allowed', '❌ Keys unique'],
    ['Memory usage', 'Higher', 'Lower', 'Moderate'],
    ['Use case', 'Dynamic sequences', 'Fixed data', 'Labeled data']
]

print("Feature Comparison:")
for row in comparison_data:
    print(f"{row[0]:<20} {row[1]:<15} {row[2]:<15} {row[3]}")

# Quick decision guide
print("\n🎯 Quick Decision Guide:")
print("• Need ordered collection that changes? → LIST")
• Need fixed data that never changes? → TUPLE")
• Need labeled data with key lookup? → DICTIONARY")
• Need composite keys for lookup? → Dictionary with TUPLE keys")
• Need to protect data integrity? → TUPLE")
````

````python
# Real-World Healthcare Data Example
print("\n=== REAL-WORLD EXAMPLE: HEALTHCARE DATA SYSTEM ===")

# Hospital patient management system using all three structures
hospital_db = {
    'patients': [  # LIST of patient dictionaries
        {
            'id': 'P001',
            'name': 'Alice Johnson',
            'demographics': ('1985-03-20', 'F', 'O+'),  # TUPLE for fixed data
            'vitals_history': [  # LIST for changing data
                {'date': '2024-04-20', 'bp': '120/80', 'hr': 72},
                {'date': '2024-04-21', 'bp': '118/78', 'hr': 70},
                {'date': '2024-04-22', 'bp': '122/82', 'hr': 74}
            ],
            'medications': ['Lisinopril', 'Metformin']
        },
        {
            'id': 'P002',
            'name': 'Bob Smith',
            'demographics': ('1978-11-12', 'M', 'A+'),
            'vitals_history': [
                {'date': '2024-04-20', 'bp': '135/85', 'hr': 78}
            ],
            'medications': ['Atorvastatin']
        }
    ],
    'departments': ('ER', 'ICU', 'General', 'Surgery'),  # TUPLE - fixed departments
    'stats': {  # DICTIONARY for labeled statistics
        'total_patients': 2,
        'avg_age': 42.5,
        'critical_cases': 0
    }
}

print("Hospital Database Structure:")
print(f"Total patients: {hospital_db['stats']['total_patients']}")
print(f"Departments: {hospital_db['departments']}")

print("\nPatient Details:")
for patient in hospital_db['patients']:
    dob, gender, blood = patient['demographics']
    print(f"  {patient['name']} ({patient['id']}) - {gender}, Blood: {blood}")
    print(f"    Medications: {patient['medications']}")
    print(f"    Latest vitals: {patient['vitals_history'][-1]}")
    print()

print("✅ Demonstrates effective use of all three collection types!")
````

## 5. Common Operations Summary

### Essential Operations Reference

#### List Operations:
```python
# Creating
my_list = [1, 2, 3]

# Accessing
first = my_list[0]          # Index 0
last = my_list[-1]          # Last item
slice = my_list[1:3]        # Items 1-2

# Modifying
my_list[0] = 10             # Change item
my_list.append(4)           # Add to end
my_list.insert(0, 0)        # Insert at position
my_list.remove(2)           # Remove value
item = my_list.pop()        # Remove and return last

# Information
length = len(my_list)
exists = 3 in my_list
count = my_list.count(2)
index = my_list.index(3)
```

#### Tuple Operations:
```python
# Creating
my_tuple = (1, 2, 3)
single = (1,)               # Note the comma!

# Accessing (same as lists)
first = my_tuple[0]
last = my_tuple[-1]

# Unpacking
a, b, c = my_tuple
first, *rest, last = my_tuple

# Information
length = len(my_tuple)
exists = 2 in my_tuple
count = my_tuple.count(1)
index = my_tuple.index(2)
```

#### Dictionary Operations:
```python
# Creating
my_dict = {'key': 'value'}

# Accessing
value = my_dict['key']      # KeyError if not found
value = my_dict.get('key')  # Returns None if not found
value = my_dict.get('key', 'default')  # Returns default if not found

# Modifying
my_dict['new_key'] = 'value'    # Add new
my_dict['key'] = 'new_value'    # Update existing
my_dict.update({'a': 1, 'b': 2}) # Add multiple
value = my_dict.pop('key')      # Remove and return
item = my_dict.popitem()        # Remove arbitrary item
del my_dict['key']              # Delete key

# Information
keys = my_dict.keys()
values = my_dict.values()
items = my_dict.items()
length = len(my_dict)
exists = 'key' in my_dict
```
