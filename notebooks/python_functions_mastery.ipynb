## 1. Defining Functions - Creating Reusable Code Blocks

### What is a Function?
A **function** is a named block of code that performs a specific task. Functions help you organize code, reduce repetition, and make programs modular and maintainable.

### Why Use Functions?
- ✅ **Reusability**: Write once, use many times
- ✅ **Organization**: Break complex problems into smaller parts
- ✅ **Readability**: Code is easier to understand with named operations
- ✅ **Maintainability**: Changes only need to be made in one place
- ✅ **Testing**: Functions can be tested individually
- ✅ **Abstraction**: Hide complex logic behind simple interfaces

### Function Definition Syntax:
```python
def function_name():
    """
    Docstring explaining what the function does.
    """
    # Function body (indented code)
    # Perform some operations
    return result  # Optional
```

### Key Concepts:
- ✅ **def keyword**: Starts function definition
- ✅ **Function name**: Should be descriptive (verb_noun format)
- ✅ **Parentheses ()**: Hold parameters (can be empty)
- ✅ **Colon :**: Ends function signature
- ✅ **Indentation**: Function body must be indented (4 spaces)
- ✅ **Docstring**: Explains what the function does
- ✅ **return statement**: Sends result back to caller (optional)

````python
# Simple Function Definitions
print("=== BASIC FUNCTION DEFINITIONS ===")

# Function 1: No parameters, no return
def greet_patient():
    """
    Prints a greeting message for a patient.
    This function takes no arguments and returns nothing.
    """
    print("👋 Welcome to MediPredict Healthcare System!")
    print("🩺 Please check in at the reception.")

# Function 2: Simple calculation, returns value
def calculate_bmi(weight_kg, height_m):
    """
    Calculate Body Mass Index (BMI).
    
    Args:
        weight_kg: Weight in kilograms
        height_m: Height in meters
    
    Returns:
        BMI value as float
    """
    bmi = weight_kg / (height_m ** 2)
    return bmi

# Function 3: Multiple operations with return
def get_vital_signs_status(heart_rate, blood_pressure):
    """
    Assess vital signs and return status.
    
    Args:
        heart_rate: Heart rate in beats per minute
        blood_pressure: Blood pressure as "systolic/diastolic"
    
    Returns:
        Status string: "Normal", "Elevated", or "Critical"
    """
    # Parse blood pressure
    systolic, diastolic = map(int, blood_pressure.split('/'))
    
    # Check heart rate
    hr_normal = 60 <= heart_rate <= 100
    
    # Check blood pressure
    bp_normal = systolic < 120 and diastolic < 80
    
    if hr_normal and bp_normal:
        return "Normal"
    elif heart_rate > 120 or systolic > 140:
        return "Critical"
    else:
        return "Elevated"

print("✅ Functions defined successfully!")
print(f"  - greet_patient(): {type(greet_patient)}")
print(f"  - calculate_bmi(): {type(calculate_bmi)}")
print(f"  - get_vital_signs_status(): {type(get_vital_signs_status)}")
````

````python
# Healthcare Data Processing Functions
print("\n=== HEALTHCARE DATA PROCESSING FUNCTIONS ===")

def validate_patient_age(age):
    """
    Validate if age is within acceptable range.
    
    Args:
        age: Patient age in years
    
    Returns:
        True if valid, False otherwise
    """
    return isinstance(age, (int, float)) and 0 <= age <= 120

def format_patient_name(first_name, last_name):
    """
    Format patient name with proper capitalization.
    
    Args:
        first_name: Patient's first name
        last_name: Patient's last name
    
    Returns:
        Formatted full name
    """
    formatted_first = first_name.strip().title()
    formatted_last = last_name.strip().title()
    return f"{formatted_first} {formatted_last}"

def classify_blood_pressure(systolic, diastolic):
    """
    Classify blood pressure reading.
    
    Args:
        systolic: Systolic pressure in mmHg
        diastolic: Diastolic pressure in mmHg
    
    Returns:
        Classification string
    """
    if systolic < 120 and diastolic < 80:
        return "Normal"
    elif 120 <= systolic <= 129 and diastolic < 80:
        return "Elevated"
    elif 130 <= systolic <= 139 or 80 <= diastolic <= 89:
        return "Stage 1 Hypertension"
    elif systolic >= 140 or diastolic >= 90:
        return "Stage 2 Hypertension"
    else:
        return "Unknown"

def calculate_age_group(age):
    """
    Categorize patient into age group.
    
    Args:
        age: Patient age in years
    
    Returns:
        Age group string
    """
    if age < 0:
        return "Invalid"
    elif age < 13:
        return "Child"
    elif age < 20:
        return "Teenager"
    elif age < 65:
        return "Adult"
    else:
        return "Senior"

# Test the functions
print("Testing function definitions:")
print(f"  validate_patient_age(25): {validate_patient_age(25)}")
print(f"  validate_patient_age(-5): {validate_patient_age(-5)}")
print(f"  format_patient_name('john', 'doe'): {format_patient_name('john', 'doe')}")
print(f"  classify_blood_pressure(120, 80): {classify_blood_pressure(120, 80)}")
print(f"  calculate_age_group(45): {calculate_age_group(45)}")
````

## 2. Calling Functions - Executing Reusable Code

### What is Calling a Function?
**Calling** (or invoking) a function means executing the code inside it. When you call a function, control jumps to the function body, executes the code, and then returns to where it was called from.

### Function Call Syntax:
```python
# Call function with no arguments
function_name()

# Call function with arguments
function_name(arg1, arg2)

# Call function and capture return value
result = function_name(arg1)
```

### Execution Flow:
1. **Call**: Program jumps to function definition
2. **Execute**: Function body runs with provided arguments
3. **Return**: Function completes and control returns to caller
4. **Continue**: Program continues from after the function call

````python
# Function Calling Examples
print("=== CALLING FUNCTIONS ===")

# Call function with no arguments, no return
print("1. Calling greet_patient():")
greet_patient()

# Call function with arguments and capture return
print("\n2. Calling calculate_bmi():")
patient_weight = 70  # kg
patient_height = 1.75  # meters

patient_bmi = calculate_bmi(patient_weight, patient_height)
print(f"  Weight: {patient_weight} kg")
print(f"  Height: {patient_height} m")
print(f"  BMI: {patient_bmi:.2f}")

# Call function multiple times with different values
print("\n3. Calling calculate_bmi() for multiple patients:")
patients_data = [
    (65, 1.70),   # Patient 1
    (80, 1.80),   # Patient 2
    (55, 1.60),   # Patient 3
]

for i, (weight, height) in enumerate(patients_data, 1):
    bmi = calculate_bmi(weight, height)
    print(f"  Patient {i}: BMI = {bmi:.2f}")

# Call function and use result in condition
print("\n4. Using function result in condition:")
test_bmi = calculate_bmi(85, 1.75)

if test_bmi > 30:
    category = "Obese"
elif test_bmi > 25:
    category = "Overweight"
elif test_bmi > 18.5:
    category = "Normal"
else:
    category = "Underweight"

print(f"  BMI: {test_bmi:.2f} -> Category: {category}")
````

````python
# Function Call Execution Flow Demonstration
print("\n=== FUNCTION EXECUTION FLOW ===")

def demonstrate_execution_flow():
    """
    This function demonstrates execution flow with print statements.
    """
    print("    [Inside function] Starting function execution")
    result = 42
    print(f"    [Inside function] Calculated result: {result}")
    print("    [Inside function] About to return")
    return result

print("1. Program starts")
print("2. About to call function")

returned_value = demonstrate_execution_flow()

print("3. Back from function call")
print(f"4. Returned value: {returned_value}")
print("5. Program continues...")

print("\n=== CHAINING FUNCTION CALLS ===")

def get_raw_data():
    """Simulate fetching raw patient data."""
    return [98.6, 99.1, -999, 101.5, 97.8]

def clean_data(data):
    """Remove invalid temperature readings."""
    return [x for x in data if x > 0]

def calculate_average(data):
    """Calculate average of cleaned data."""
    return sum(data) / len(data) if data else 0

# Chain function calls
raw = get_raw_data()
cleaned = clean_data(raw)
average = calculate_average(cleaned)

print(f"Raw data: {raw}")
print(f"Cleaned data: {cleaned}")
print(f"Average temperature: {average:.2f}°F")

# Or chain in one line (less readable)
direct_average = calculate_average(clean_data(get_raw_data()))
print(f"Direct calculation: {direct_average:.2f}°F")
````

## 3. Parameters and Arguments - Passing Data to Functions

### What are Parameters and Arguments?
- **Parameters**: Variables listed in the function definition (placeholders)
- **Arguments**: Actual values passed to the function when calling it

### Types of Arguments:
- ✅ **Positional Arguments**: Matched by position/order
- ✅ **Keyword Arguments**: Matched by parameter name
- ✅ **Default Values**: Parameters with predefined values
- ✅ **Variable Arguments**: *args for multiple positional arguments
- ✅ **Variable Keywords**: **kwargs for multiple keyword arguments

### Best Practices:
- ✅ Use clear, descriptive parameter names
- ✅ Provide default values for optional parameters
- ✅ Use type hints when possible (Python 3.5+)
- ✅ Keep number of parameters reasonable (3-4 max ideally)

````python
# Parameters and Arguments Examples
print("=== PARAMETERS AND ARGUMENTS ===")

# Function with required positional parameters
def calculate_dosage(weight, age, medication):
    """
    Calculate medication dosage based on patient info.
    
    Args:
        weight: Patient weight in kg
        age: Patient age in years
        medication: Name of medication
    """
    base_dose = 10  # mg per kg
    
    # Adjust for age
    if age < 12:
        adjustment = 0.5  # 50% for children
    elif age > 65:
        adjustment = 0.75  # 75% for seniors
    else:
        adjustment = 1.0  # Full dose for adults
    
    dosage = weight * base_dose * adjustment
    return {
        'medication': medication,
        'dosage_mg': round(dosage, 1),
        'adjustment': adjustment,
        'weight': weight,
        'age': age
    }

# Calling with positional arguments (order matters!)
print("1. Positional arguments:")
dose1 = calculate_dosage(70, 45, 'Aspirin')
print(f"  Adult (70kg, 45y): {dose1}")

dose2 = calculate_dosage(25, 8, 'Amoxicillin')
print(f"  Child (25kg, 8y): {dose2}")

# Calling with keyword arguments (order doesn't matter)
print("\n2. Keyword arguments:")
dose3 = calculate_dosage(age=70, weight=65, medication='Metformin')
print(f"  Senior (65kg, 70y): {dose3}")

# Mixing positional and keyword (positional must come first)
print("\n3. Mixed arguments:")
dose4 = calculate_dosage(80, medication='Lisinopril', age=55)
print(f"  Adult (80kg, 55y): {dose4}")
````

````python
# Default Parameter Values
print("\n=== DEFAULT PARAMETER VALUES ===")

def record_vitals(patient_id, temperature=98.6, heart_rate=72, bp_systolic=120, bp_diastolic=80):
    """
    Record patient vital signs with default normal values.
    
    Args:
        patient_id: Unique patient identifier (required)
        temperature: Body temperature in Fahrenheit (default: 98.6)
        heart_rate: Heart rate in bpm (default: 72)
        bp_systolic: Systolic BP in mmHg (default: 120)
        bp_diastolic: Diastolic BP in mmHg (default: 80)
    """
    return {
        'patient_id': patient_id,
        'temperature': temperature,
        'heart_rate': heart_rate,
        'blood_pressure': f"{bp_systolic}/{bp_diastolic}",
        'recorded_at': '2024-04-30 10:00'
    }

print("1. Using all defaults:")
vitals1 = record_vitals('P001')
print(f"  {vitals1}")

print("\n2. Overriding some defaults:")
vitals2 = record_vitals('P002', temperature=101.2, heart_rate=95)
print(f"  {vitals2}")

print("\n3. Overriding all values:")
vitals3 = record_vitals('P003', 99.1, 88, 135, 85)
print(f"  {vitals3}")

print("\n4. Selective overrides with keywords:")
vitals4 = record_vitals('P004', bp_systolic=140, bp_diastolic=90)
print(f"  {vitals4}")

# Important: Default values are evaluated once!
print("\n=== IMPORTANT: MUTABLE DEFAULT VALUES ===")

# BAD - Don't use mutable default values!
def bad_add_medication(med_list=[]):
    med_list.append('Aspirin')
    return med_list

# GOOD - Use None and check inside function
def good_add_medication(med_list=None):
    if med_list is None:
        med_list = []
    med_list.append('Aspirin')
    return med_list

print("Testing bad function:")
result1 = bad_add_medication()
result2 = bad_add_medication()
print(f"  Call 1: {result1}")
print(f"  Call 2: {result2}")  # Notice it accumulates!

print("\nTesting good function:")
result3 = good_add_medication()
result4 = good_add_medication()
print(f"  Call 1: {result3}")
print(f"  Call 2: {result4}")  # Each call is independent
````

````python
# Variable Arguments (*args and **kwargs)
print("\n=== VARIABLE ARGUMENTS ===")

# *args - Variable number of positional arguments
def calculate_statistics(*readings):
    """
    Calculate statistics for variable number of readings.
    
    Args:
        *readings: Variable number of numeric readings
    
    Returns:
        Dictionary with count, min, max, avg
    """
    if not readings:
        return {'count': 0, 'min': None, 'max': None, 'avg': None}
    
    return {
        'count': len(readings),
        'min': min(readings),
        'max': max(readings),
        'avg': sum(readings) / len(readings)
    }

print("1. Using *args - variable readings:")
stats1 = calculate_statistics(98.6, 99.1, 97.8, 101.5)
print(f"  Temperatures: {stats1}")

stats2 = calculate_statistics(120, 118, 122, 125, 119, 121)
print(f"  Blood pressures: {stats2}")

# **kwargs - Variable number of keyword arguments
def create_patient_record(patient_id, **patient_data):
    """
    Create a patient record with required ID and optional data fields.
    
    Args:
        patient_id: Required unique identifier
        **patient_data: Optional keyword arguments for patient data
    
    Returns:
        Complete patient record dictionary
    """
    record = {'patient_id': patient_id}
    record.update(patient_data)
    record['created_at'] = '2024-04-30'
    return record

print("\n2. Using **kwargs - flexible patient data:")

patient1 = create_patient_record('P001',
                                name='Alice Johnson',
                                age=34,
                                blood_type='O+')
print(f"  Patient 1: {patient1}")

patient2 = create_patient_record('P002',
                                name='Bob Smith',
                                age=45,
                                weight=80,
                                height=175,
                                allergies=['Penicillin'],
                                medications=['Metformin'])
print(f"  Patient 2: {patient2}")

# Combining all types
def flexible_function(required_param, optional_param='default', *args, **kwargs):
    """
    Demonstrates all parameter types together.
    """
    return {
        'required': required_param,
        'optional': optional_param,
        'args': args,
        'kwargs': kwargs
    }

print("\n3. Combining all parameter types:")
result = flexible_function('required_value',
                        'custom_optional',
                        'extra1', 'extra2', 'extra3',
                        key1='value1',
                        key2='value2')
print(f"  Result: {result}")
````

## 4. Function Scope - Variable Visibility

### What is Scope?
**Scope** refers to where variables can be accessed in your code. Understanding scope helps prevent bugs and makes code more predictable.

### Types of Scope:
- ✅ **Local Scope**: Variables created inside a function (only accessible there)
- ✅ **Global Scope**: Variables created outside functions (accessible everywhere)
- ✅ **LEGB Rule**: Python looks for variables in this order: Local, Enclosing, Global, Built-in

### Best Practices:
- ✅ **Prefer local variables** - Keep functions self-contained
- ✅ **Minimize global variables** - They can cause unexpected side effects
- ✅ **Pass data explicitly** - Use parameters and return values
- ✅ **Use global keyword sparingly** - Only when truly necessary

````python
# Function Scope Examples
print("=== FUNCTION SCOPE ===")

# Local scope - variables inside functions
def demonstrate_local_scope():
    """
    Shows that variables inside functions are local.
    """
    local_variable = "I am local"
    print(f"  Inside function: {local_variable}")
    return local_variable

print("1. Local scope demonstration:")
result = demonstrate_local_scope()
print(f"  Returned value: {result}")

# This would cause an error - local_variable doesn't exist here
# print(f"Outside function: {local_variable}")  # NameError!
print("  ✓ local_variable is not accessible outside the function")

# Global scope - variables outside functions
global_hospital_name = "City General Hospital"

def use_global_variable():
    """
    Access global variable inside function.
    """
    print(f"  Inside function: {global_hospital_name}")
    return f"Welcome to {global_hospital_name}"

print("\n2. Global scope demonstration:")
print(f"  Outside function: {global_hospital_name}")
welcome_msg = use_global_variable()
print(f"  Function returned: {welcome_msg}")

# Shadowing - local variable with same name as global
patient_count = 100  # Global

def count_patients():
    patient_count = 5  # Local - shadows global
    print(f"  Inside function (local): {patient_count}")
    return patient_count

print("\n3. Variable shadowing:")
print(f"  Before function call (global): {patient_count}")
local_count = count_patients()
print(f"  After function call (global unchanged): {patient_count}")
print(f"  Function returned (local value): {local_count}")
````

````python
# Modifying Global Variables
print("\n=== MODIFYING GLOBAL VARIABLES ===")

# Without global keyword (creates local variable)
total_patients_processed = 0

def process_patient_bad(patient_name):
    """
    WRONG: This creates a local variable, doesn't modify global.
    """
    total_patients_processed = total_patients_processed + 1  # UnboundLocalError!
    print(f"  Processed {patient_name}")

# With global keyword (modifies global variable)
total_patients_processed = 0

def process_patient_good(patient_name):
    """
    CORRECT: Uses global keyword to modify global variable.
    """
    global total_patients_processed
    total_patients_processed = total_patients_processed + 1
    print(f"  Processed {patient_name} (Total: {total_patients_processed})")

print("1. Better approach - pass and return values:")

def process_patient_better(current_count, patient_name):
    """
    BEST: No global variables, pure function.
    """
    new_count = current_count + 1
    print(f"  Processed {patient_name} (Total: {new_count})")
    return new_count

# Demonstration
count = 0
count = process_patient_better(count, "Alice")
count = process_patient_better(count, "Bob")
count = process_patient_better(count, "Carol")
print(f"\nFinal count: {count}")

print("\n2. When to use global (rarely needed):")

# Configuration or constants
HOSPITAL_NAME = "MediPredict General"
DEFAULT_TIMEOUT = 30

def get_hospital_info():
    """Read-only access to global constants is fine."""
    return {
        'name': HOSPITAL_NAME,
        'timeout': DEFAULT_TIMEOUT
    }

info = get_hospital_info()
print(f"  Hospital info: {info}")

print("\n3. Mutable global objects (avoid modifying):")

# Global list - DON'T modify inside functions
all_patients = ['Alice', 'Bob']

def bad_add_patient(patient):
    """MODIFYING global list - BAD PRACTICE!"""
    all_patients.append(patient)

def good_add_patient(patient_list, patient):
    """Return new list instead of modifying - GOOD PRACTICE!"""
    return patient_list + [patient]

print(f"  Original list: {all_patients}")

# Bad approach
bad_add_patient('Carol')
print(f"  After bad_add: {all_patients}")  # Original modified!

# Good approach
new_list = good_add_patient(all_patients, 'David')
print(f"  Original unchanged: {all_patients}")
print(f"  New list returned: {new_list}")
````

## 5. Practical Function Design

### Function Design Best Practices:
- ✅ **Single Responsibility**: Each function does one thing well
- ✅ **Descriptive Names**: Function name describes what it does
- ✅ **Docstrings**: Explain purpose, parameters, and return values
- ✅ **Input Validation**: Check parameters are valid
- ✅ **Return Values**: Return useful data, not just print
- ✅ **Pure Functions**: Same input → Same output, no side effects
- ✅ **Keep Short**: Functions should fit on one screen (~20-30 lines)

````python
# Practical Healthcare Functions
print("=== PRACTICAL HEALTHCARE FUNCTIONS ===")

def validate_patient_data(name, age, weight, height):
    """
    Validate patient data and return list of errors.
    
    Args:
        name: Patient name (non-empty string)
        age: Age in years (0-120)
        weight: Weight in kg (positive)
        height: Height in cm (positive)
    
    Returns:
        List of error messages (empty if valid)
    """
    errors = []
    
    if not name or not isinstance(name, str):
        errors.append('Name must be a non-empty string')
    
    if not isinstance(age, (int, float)) or age < 0 or age > 120:
        errors.append('Age must be between 0 and 120')
    
    if not isinstance(weight, (int, float)) or weight <= 0:
        errors.append('Weight must be positive')
    
    if not isinstance(height, (int, float)) or height <= 0:
        errors.append('Height must be positive')
    
    return errors

def calculate_risk_score(age, smoker, family_history):
    """
    Calculate cardiovascular risk score.
    
    Args:
        age: Patient age
        smoker: True if patient smokes
        family_history: True if family has heart disease
    
    Returns:
        Risk score (0-10) and risk level
    """
    score = 0
    
    # Age factor
    if age >= 65:
        score += 3
    elif age >= 45:
        score += 2
    elif age >= 30:
        score += 1
    
    # Smoking factor
    if smoker:
        score += 3
    
    # Family history factor
    if family_history:
        score += 2
    
    # Determine risk level
    if score >= 6:
        level = "High"
    elif score >= 3:
        level = "Moderate"
    else:
        level = "Low"
    
    return {'score': score, 'level': level}

def format_medical_record(patient_data):
    """
    Format patient data as readable medical record.
    
    Args:
        patient_data: Dictionary with patient information
    
    Returns:
        Formatted string
    """
    name = patient_data.get('name', 'Unknown')
    age = patient_data.get('age', 'N/A')
    bp = patient_data.get('blood_pressure', 'N/A')
    hr = patient_data.get('heart_rate', 'N/A')
    
    return f"""
MEDICAL RECORD
==============
Patient: {name}
Age: {age} years
Blood Pressure: {bp}
Heart Rate: {hr} bpm
"""

# Test the practical functions
print("1. Testing validate_patient_data():")
errors1 = validate_patient_data('John Doe', 45, 75, 180)
print(f"  Valid patient: {errors1}")

errors2 = validate_patient_data('', -5, -10, 0)
print(f"  Invalid patient errors: {errors2}")

print("\n2. Testing calculate_risk_score():")
risk1 = calculate_risk_score(55, True, True)
print(f"  55yo smoker with family history: {risk1}")

risk2 = calculate_risk_score(25, False, False)
print(f"  25yo non-smoker: {risk2}")

print("\n3. Testing format_medical_record():")
record = format_medical_record({
    'name': 'Jane Smith',
    'age': 34,
    'blood_pressure': '120/80',
    'heart_rate': 72
})
print(record)
````

````python
# Building a Complete Workflow with Functions
print("\n=== COMPLETE HEALTHCARE WORKFLOW ===")

def process_patient_batch(raw_patient_data):
    """
    Complete workflow: validate, process, and report on patient data.
    """
    valid_patients = []
    invalid_patients = []
    risk_summaries = []
    
    for patient in raw_patient_data:
        # Step 1: Validate
        errors = validate_patient_data(
            patient.get('name'),
            patient.get('age'),
            patient.get('weight'),
            patient.get('height')
        )
        
        if errors:
            invalid_patients.append({
                'patient': patient,
                'errors': errors
            })
            continue
        
        # Step 2: Calculate risk
        risk = calculate_risk_score(
            patient.get('age', 0),
            patient.get('smoker', False),
            patient.get('family_history', False)
        )
        
        # Step 3: Add to valid patients with risk info
        patient_with_risk = {**patient, **risk}
        valid_patients.append(patient_with_risk)
        risk_summaries.append(risk['level'])
    
    # Generate summary
    summary = {
        'total_processed': len(raw_patient_data),
        'valid': len(valid_patients),
        'invalid': len(invalid_patients),
        'risk_distribution': {
            'low': risk_summaries.count('Low'),
            'moderate': risk_summaries.count('Moderate'),
            'high': risk_summaries.count('High')
        }
    }
    
    return {
        'valid_patients': valid_patients,
        'invalid_patients': invalid_patients,
        'summary': summary
    }

# Sample data
raw_data = [
    {'name': 'Alice Johnson', 'age': 34, 'weight': 65, 'height': 165, 'smoker': False, 'family_history': False},
    {'name': 'Bob Smith', 'age': 58, 'weight': 85, 'height': 175, 'smoker': True, 'family_history': True},
    {'name': '', 'age': 25, 'weight': 60, 'height': 160, 'smoker': False, 'family_history': False},  # Invalid
    {'name': 'Carol White', 'age': 45, 'weight': 70, 'height': 170, 'smoker': False, 'family_history': True},
    {'name': 'David Brown', 'age': 67, 'weight': 78, 'height': 180, 'smoker': True, 'family_history': True},
]

result = process_patient_batch(raw_data)

print(f"Processed {result['summary']['total_processed']} patients")
print(f"✅ Valid: {result['summary']['valid']}")
print(f"❌ Invalid: {result['summary']['invalid']}")
print(f"\nRisk Distribution:")
for level, count in result['summary']['risk_distribution'].items():
    print(f"  {level}: {count}")

print(f"\nInvalid Records:")
for invalid in result['invalid_patients']:
    print(f"  {invalid['patient'].get('name', 'Unknown')}: {invalid['errors']}")
````
