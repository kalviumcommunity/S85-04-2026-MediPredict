```python
# Verify Python version and basic functionality
import sys
print(f"Python version: {sys.version}")
print("✅ Python is working correctly!")
```

```python
# Verify data science packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"Matplotlib version: {plt.__version__}")
print("✅ All data science packages imported successfully!")
```

```python
# Test basic data operations
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
})

print("Sample DataFrame:")
print(data)
print(f"\nData shape: {data.shape}")
print(f"Mean of column A: {data['A'].mean()}")
print("✅ Data operations working correctly!")
```

```python
# Test simple plot
plt.figure(figsize=(8, 4))
plt.plot(data['A'], data['B'], 'bo-', label='Sample Data')
plt.xlabel('Column A')
plt.ylabel('Column B')
plt.title('Sample Plot - Verification Test')
plt.legend()
plt.grid(True)
plt.show()
print("✅ Matplotlib plotting working correctly!")
```
