## 1. Running Cells and Understanding Execution Order

### Key Concepts:
- **Kernel State**: Variables persist until kernel restart
- **Execution Order**: Matters more than cell order
- **Hidden State**: Variables can affect later cells

Let's demonstrate this with variable tracking:

```python
# Cell 1: Initialize variables
print("📍 Cell 1: Initializing variables")

# Create some variables
counter = 0
data_list = []
execution_log = []

print(f"counter = {counter}")
print(f"data_list = {data_list}")
print(f"execution_log = {execution_log}")
print("✅ Variables initialized successfully!")
```

```python
# Cell 2: Modify variables (demonstrates state persistence)
print("📍 Cell 2: Modifying variables")

# Update the variables from Cell 1
counter += 1
data_list.append(f"item_{counter}")
execution_log.append(f"Execution {counter}")

print(f"counter = {counter}")
print(f"data_list = {data_list}")
print(f"execution_log = {execution_log}")
print("✅ Variables modified - state persists!")
```

```python
# Cell 3: Demonstrate execution order importance
print("📍 Cell 3: Execution order demonstration")

# Try to use a variable that should exist from previous cells
try:
    print(f"Current counter value: {counter}")
    print(f"Data list length: {len(data_list)}")
    print(f"Last execution: {execution_log[-1] if execution_log else 'None'}")
    print("✅ All variables accessible - proper execution order!")
except NameError as e:
    print(f"❌ Error: {e}")
    print("This happens if cells run out of order or after kernel restart!")
```

## 2. Restarting the Kernel

### What Kernel Restart Does:
- ✅ **Clears all variables** from memory
- ✅ **Resets import state** (need to re-import libraries)
- ✅ **Clears execution history** (cell numbers reset)
- ✅ **Stops all running operations**

### When to Restart:
- 🔄 **Before final submission** (ensure reproducibility)
- 🔄 **When variables are corrupted** or have unexpected values
- 🔄 **After major code changes** that affect variable structure
- 🔄 **When notebook behaves inconsistently**

**👆 ACTION**: Now restart the kernel (Kernel → Restart & Clear Output) and run the next cell to see the effect!

```python
# Cell 4: Test after kernel restart
print("📍 Cell 4: Testing after kernel restart")

# Try to access variables from before restart
try:
    print(f"counter = {counter}")
    print("❌ Variables still exist - kernel not restarted!")
except NameError as e:
    print(f"✅ Expected error: {e}")
    print("✅ Kernel was successfully restarted - all variables cleared!")

# Re-initialize to continue demonstration
print("\n🔄 Re-initializing variables after restart...")
counter = 0
data_list = []
execution_log = []
print("✅ Variables re-initialized - ready to continue!")
```

## 3. Interrupting Long-Running Execution

### What Interrupt Does:
- ⏹️ **Stops current cell execution** immediately
- ✅ **Preserves kernel state** (variables remain)
- ✅ **Keeps notebook responsive** for other operations
- ⚠️ **May leave operations in incomplete state**

### When to Interrupt:
- 🛑 **Infinite loops** or stuck computations
- 🛑 **Very long operations** that you want to stop
- 🛑 **Accidentally large computations** (e.g., huge loops)

**👆 ACTION**: Run the next cell, then interrupt it (■ button or Kernel → Interrupt)!

```python
# Cell 5: Long-running operation to interrupt
print("📍 Cell 5: Starting long-running operation...")
print("👆 INTERRUPT THIS CELL to test interrupt functionality!")

# Simulate a long-running operation
import time
counter = 0

print("Starting countdown...")
for i in range(100):  # This will run for a while
    print(f"Count: {i}/100")
    time.sleep(0.1)  # Sleep to make it take longer
    counter += 1

print("✅ Operation completed without interruption!")
```

```python
# Cell 6: Test interrupt effect
print("📍 Cell 6: Testing interrupt effect")

# Check if the counter variable exists and its value
try:
    print(f"Counter value after interrupt: {counter}")
    if counter < 100:
        print("✅ Interrupt worked - operation stopped early!")
        print("✅ Kernel state preserved - variable still exists!")
    else:
        print("ℹ️ Operation completed normally or counter not affected")
except NameError:
    print("ℹ️ Counter variable not found - may need to run previous cell first")

print("\n📝 Interrupt Summary:")
print("- ✅ Stops current execution immediately")
print("- ✅ Preserves existing variables")
print("- ✅ Notebook remains responsive")
```

## 4. When to Restart vs Interrupt - Decision Guide

### 🛑 Use INTERRUPT When:
- **Current cell is stuck** but previous work is fine
- **Long-running operation** needs to be stopped
- **Infinite loop** accidentally created
- **Want to preserve** current variable state

### 🔄 Use RESTART When:
- **Variables are corrupted** or have wrong values
- **Notebook behaves inconsistently** (cells failing unexpectedly)
- **Before final submission** (ensure reproducibility)
- **Import issues** or library conflicts
- **Memory issues** or kernel is sluggish

### ⚡ Quick Reference:
| Situation | Action | Reason |
|-----------|--------|--------|
| Stuck in infinite loop | 🛑 Interrupt | Stop execution, keep variables |
| Variables have wrong values | 🔄 Restart | Clear corrupted state |
| Before sharing/submission | 🔄 Restart | Ensure reproducibility |
| Long computation needs stopping | 🛑 Interrupt | Preserve work done |
| Import errors after code changes | 🔄 Restart | Reset import state |
